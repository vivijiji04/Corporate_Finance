### Modern Firm

**The Four Types of Firms**
*   Sole Proprietorship
*   Partnership
*   Limited Liability Company
*   Corporation

**Firms differ in several factors. Those that are most important to financial decision-making are:**
*   Taxation
*   Degree of control
*   Owners' liability
*   Ease of transferring ownership
*   Ability to raise additional funds
*   Longevity of the business

---


### Sole Proprietorship
*   Business is owned and run by one person
*   Typically has few, if any, employees
*   Easy to create
*   Disadvantages
    *   No separation between the firm and the owner
    *   Unlimited personal liability
    *   Limited life
    *   Limited access to additional funds.

---

### Partnership
*   More than one owner
*   All partners have unlimited personal liability
*   The partnership ends with the death or withdrawal of any single partner.
*   Limited Partnership
    *   General Partners
    *   Limited Partners

### Limited Liability Companies (LLC)
*   A hybrid of partnership and corporation
*   Usually, a legal entity separate from its owners
*   Usually operate and be taxed like a partnership but retain limited liability for owners
    *   Double taxation
    
---

### Corporation
*   A legal entity separate from its owners
    *   The ability to enter into contracts, own assets, and borrow money
    *   Solely responsible for its own obligations
*   Ownership
    *   Represented by shares of stock
    *   Sum of all ownership value is called equity.
    *   There is no limit to the number of shareholders and, thus, the amount of funds a company can raise by selling stock.
    *   Owners are entitled to dividend payments.
*   Starting a corporation is more complicated than starting a proprietorship or partnership.
    *   The incorporators must prepare articles of incorporation and a set of bylaws.
        *   Name of the corporation.
        *   Intended life of the corporation (may be forever).
        *   Business purpose.
        *   Number of shares of stock that the corporation is authorized to issue, with a statement of limitations and rights of different classes of shares.
        *   Nature of the rights granted to shareholders.
        *   Number of members of the initial board of directors.

---

### Corporation: Three sets of distinct interests
*   The shareholders (owners)
    *   Control the corporation's direction, policies, and activities.
    *   Elect a board of directors, who in turn select top management.
*   The directors (board)
    *   Represent the interest of shareholders
    *   Select and oversee the top management.
*   The corporation officers (top management)
    *   Manage the operations of the corporation in the best interest of the shareholders.
*   In closely held corporations with few shareholders, there may be a large overlap among the shareholders, the directors, and the top management.


---

### Corporation: Main advantages over proprietorships and partnerships
*   Limited liability
    *   The shareholders' liability is limited to the amount invested in the ownership shares.
*   Ease of ownership transfer
    *   Ownership in a corporation is represented by shares of stock and can be readily transferred to new owners.
    *   There is no limit to the transferability.
*   Perpetual succession
    *   The corporation has unlimited life.
    *   The death or withdrawal of an owner does not affect the corporation's legal existence.
    *   Can continue after the original owners have withdrawn.
*   Results in enhanced ability to raise cash
*   One great disadvantage
    *   Double taxation


---

### S Corporation (S-Corp)
*   Similar provisions regarding ownership and capital generation.
*   The distinguishing features of S-Corp are related to taxation and flexibility of ownership.
*   Taxation
    *   An S-Corporation does not get charged corporate income tax for profits earned at the corporate level.
    *   All gains accrued by the business are attributed to the owners, who are then charged personal income tax.
    *   Not permitted to deduct the cost of fringe benefits provided for employee welfare such as healthcare and life insurance.
*   Flexibility of ownership
    *   Restrictions regarding the number of shareholders.
    *   Entities such as trusts, and other corporations are not entitled to stock ownership in S-Corps.
    *   Only one class of stock is permitted for distribution which means that each shareholder holds equal voting rights.
    *   Certain types of business entities, such as banks and insurance companies, are not permitted to hold S-Corp status.
*   A limited liability company taxable as a corporation may elect S-Corp status.
*   The S designation is more suited to smaller or new firms that want to escape the double taxation effect.

### Core Principles of Corporate Finance

**The Investment Principle**
*   Where should the firm invest its resources?
*   Invest in assets and projects that yield a return greater than the minimum acceptable hurdle rate.
    *   The hurdle rate should be higher for riskier projects and should reflect the financing mix used.
    *   Returns on projects should be measured based on cash flows generated and the timing of these cash flows.

**The Financing Principle**
*   Governs the mix of funding used to fund the investments.
*   Choose a financing mix that maximizes the value of the investments made and match the financing to the nature of the assets being financed.

**The Dividend Principle**
*   Answers the question of how much earnings should be reinvested back into the business and how much should be returned to the owners of the business.
*   If there are not enough investments that earn the hurdle rate, return the cash to the owners of the business.

### The Goal of Financial Management

*   Given the value maximization objective, what should the goal of financial management be?
    *   **The goal of financial management is to maximize the current value per share of the existing stock**, or
    *   **Maximize the value of the existing owners' equity.**
*   How can we achieve this goal?
    *   Achieved through making good decisions and avoiding poor decisions.
*   How do we know whether a decision is good or poor?
    *   Any decision that increases the value of a business is considered good, whereas one that reduces the firm value is considered poor.
    *   It is all about value creation vs. value destruction.

---

### The Objective of the Firm: The value of the existing owners' equity
*   The market value of shareholders' equity is the value of all owners' interest in the corporation, also referred to as the stock's market capitalization.

$$ \text{Market value of shareholders' equity} = (\text{Market price per share of stock}) \times (\text{Number of shares outstanding}) $$

*   What motivates investors?
    *   Future dividends
    *   Capital gains (increases in the value of the firm)
*   How much an investor would be willing to pay today for the future dividends?
    *   Present value of cash flows from the firm.
        *   The uncertainty associated with receiving future payments.
        *   The timing of these future payments.
        *   Compensation for tying up funds in this investment.
*   What if the firm is not publicly traded?

### The Goal of Financial Management

*   What are the shortcomings of limiting shareholder value creation to maximizing today's share price for today's shareholders?
    *   If investors knew as much about a company as its managers do, maximizing its current share price might be equivalent to maximizing value over time.
    *   Investors don't have complete information.
    *   Can companies pump up their share price in the short term?
        *   Shortchanging product development, reducing product quality, or skimping on safety.
        *   Ignoring likely future changes in regulation or consumer behavior.
    *   Can companies with a long strategic horizon create more value?
        *   May require forgoing short-term profits.
        *   Investments in research and development
*   **"Maximizing a company's collective value to current and future shareholders."**
*   Creating value for both current and future shareholders means managers should not take actions to increase today's share price if those actions will damage it down the road.

### How do financial managers create value?
*   Cash flows are the key.
    *   The fundamental principle of valuation is that the value of any financial asset is the present value of the expected cash flows.
    *   The firm must create more cash flow than it uses.
        *   Buy assets that generate more cash than they cost.
        *   Sell bonds, stocks, and other financial instruments that raise more cash than they cost.
*   The cash flows paid to bondholders and stockholders of the firm should be greater than the cash flows put into the firm by the bondholders and stockholders.
*   Return on invested capital (ROIC) vs. cost of capital

### Excess returns?
*   The value of a business can be simply stated as a function of the "excess returns" that it generates from both existing and new investments.
*   Return on assets (ROA), Return on equity (ROE), and Return on invested capital (ROIC)

$$
ROA = \frac{\text{Operating Income} * (1 - \text{tax rate})}{\text{Book Value of Total Assets}_{(t-1)}}
$$

$$
ROE = \frac{\text{Net Income}}{\text{Book Value of Equity}_{(t-1)}}
$$

$$
ROIC = \frac{\text{Operating Income} * (1 - \text{tax rate})}{\text{Book Value of Invested Capital}_{(t-1)}}
$$

*   ROIC is preferable to ROE because ROIC is unaffected by financial leverage and share buybacks whereas ROE varies based on a firm's chosen capital structure and buyback activity.
*   ROA cannot be compared to the cost of capital, since that cost is based on the cost of debt and equity invested in assets. Also, cash is a part of total assets even though operating income does not include the interest income from cash.


---

### Return on invested capital (ROIC) is a measure of a company's capital efficiency.
*   A company that generates a ROIC in excess of the cost of capital generally indicates that management has been able to deploy resources effectively.

$$
\text{Growth} = \text{ROIC} \times \text{Investment Rate}
$$

<center>or</center>

$$
\text{Investment Rate} = \frac{\text{Growth}}{\text{ROIC}}
$$

$$
\text{Cash Flow} = \text{Earnings} \times (1 - \text{Investment Rate})
$$

$$
\text{Cash Flow} = \text{Earnings} \times (1 - \frac{\text{Growth}}{\text{ROIC}})
$$

```python
import pandas as pd
import numpy as np
import ipywidgets as widgets
from IPython.display import display, clear_output, HTML as IPHTML # Renamed to avoid conflict
import matplotlib.pyplot as plt

# --- Core Calculation Logic ---
# (generate_projections function remains the same as in your last provided version)
def generate_projections(
    cost_of_capital_a, roic_a, inv_rate_a,
    cost_of_capital_b, roic_b, inv_rate_b,
    initial_earnings=100.0, n_periods=10
    ):
    data = {'Firm A': {}, 'Firm B': {}}
    roic_a_dec = float(roic_a) / 100.0
    inv_rate_a_dec = float(inv_rate_a) / 100.0
    roic_b_dec = float(roic_b) / 100.0
    inv_rate_b_dec = float(inv_rate_b) / 100.0

    growth_a_calc = roic_a_dec * inv_rate_a_dec
    growth_b_calc = roic_b_dec * inv_rate_b_dec

    inputs = {
        'Firm A': {'coc': float(cost_of_capital_a) / 100.0, 'inv_rate': inv_rate_a_dec, 'growth': growth_a_calc, 'roic': roic_a_dec},
        'Firm B': {'coc': float(cost_of_capital_b) / 100.0, 'inv_rate': inv_rate_b_dec, 'growth': growth_b_calc, 'roic': roic_b_dec}
    }

    for firm in ['Firm A', 'Firm B']:
        earnings = np.zeros(n_periods + 1)
        investment = np.zeros(n_periods + 1)
        fcf = np.zeros(n_periods + 1)
        pv_fcf = np.zeros(n_periods + 1)
        coc = inputs[firm]['coc']
        inv_rate = inputs[firm]['inv_rate']
        growth = inputs[firm]['growth']

        earnings[1] = initial_earnings
        investment[1] = earnings[1] * inv_rate
        fcf[1] = earnings[1] - investment[1]
        pv_fcf[1] = fcf[1] / (1 + coc)**1

        for t in range(2, n_periods + 1):
            earnings[t] = earnings[t-1] * (1 + growth)
            investment[t] = earnings[t] * inv_rate
            fcf[t] = earnings[t] - investment[t]
            pv_fcf[t] = fcf[t] / (1 + coc)**t

        data[firm]['Earnings'] = earnings[1:]
        data[firm]['Investment'] = investment[1:]
        data[firm]['FCF'] = fcf[1:]
        data[firm]['PV of FCF'] = pv_fcf[1:]
        # Optional: Could add ROIC and COC here if needed per period, but global spread is likely enough

    df_a = pd.DataFrame(data['Firm A'], index=range(1, n_periods + 1))
    df_b = pd.DataFrame(data['Firm B'], index=range(1, n_periods + 1))

    total_a = df_a.sum().rename('Total')
    total_b = df_b.sum().rename('Total')
    total_a['PV of FCF'] = df_a['PV of FCF'].sum()
    total_b['PV of FCF'] = df_b['PV of FCF'].sum()

    df_a = pd.concat([df_a, total_a.to_frame().T])
    df_b = pd.concat([df_b, total_b.to_frame().T])
    df_combined = pd.concat([df_a, df_b], axis=1, keys=['Firm A', 'Firm B'])
    return df_combined

# --- Widgets Setup ---
style = {'description_width': '150px'}
layout = widgets.Layout(width='300px')

coc_a = widgets.FloatText(value=10.00, step=0.25, description='Cost of Capital A (%):', style=style, layout=layout)
roic_a = widgets.FloatText(value=15.00, step=0.50, description='ROIC A (%):', style=style, layout=layout)
inv_rate_a = widgets.FloatText(value=50.00, step=0.50, description='Investment Rate A (%):', style=style, layout=layout) # Start with Inv Rate < 100%
growth_display_a = widgets.HTML(value="")
roic_coc_spread_a = widgets.HTML(value="") # For ROIC - COC spread

coc_b = widgets.FloatText(value=12.00, step=0.25, description='Cost of Capital B (%):', style=style, layout=layout)
roic_b = widgets.FloatText(value=10.00, step=0.50, description='ROIC B (%):', style=style, layout=layout) # Start with ROIC < COC
inv_rate_b = widgets.FloatText(value=70.00, step=0.50, description='Investment Rate B (%):', style=style, layout=layout)
growth_display_b = widgets.HTML(value="")
roic_coc_spread_b = widgets.HTML(value="") # For ROIC - COC spread

output_table = widgets.Output()
output_graphs = widgets.Output()
output_explanation = widgets.Output() # For explanatory text

recalculate_button = widgets.Button(
    description="Recalculate Projections",
    button_style='info',
    tooltip='Click to update table and graphs with current inputs',
    icon='refresh'
)

# --- Update Logic ---
def update_outputs_and_graphs():
    with output_explanation:
        clear_output(wait=True)
        # This text will be static once set, or could be dynamic if needed
        display(IPHTML("""
        <h4>Understanding Value Creation/Destruction:</h4>
        <ul>
            <li><b>ROIC > COC (Positive Spread):</b> The company earns more on its new investments than its cost of capital.
                Growth through such investments typically <i>creates</i> value (increases Total PV of FCF).</li>
            <li><b>ROIC < COC (Negative Spread):</b> The company earns less on its new investments than its cost of capital.
                Growth through such investments typically <i>destroys</i> value (can decrease Total PV of FCF compared to investing less or not at all).</li>
            <li><b>Observe:</b> Pay attention to the "Total PV of FCF" under different scenarios.
                Try setting ROIC < COC and see what happens to Total PV of FCF as you change the Investment Rate.
                An Investment Rate of 0% means all earnings are FCF (no growth from reinvestment).
            </li>
        </ul>
        """))

    try:
        coca_val = coc_a.value
        roica_val = roic_a.value
        inva_val = inv_rate_a.value
        cocb_val = coc_b.value
        roicb_val = roic_b.value
        invb_val = inv_rate_b.value

        implied_growth_a = (float(roica_val) / 100.0) * (float(inva_val) / 100.0)
        implied_growth_b = (float(roicb_val) / 100.0) * (float(invb_val) / 100.0)
        growth_display_a.value = f"<b>Implied Growth A: {implied_growth_a*100:.2f}%</b>"
        growth_display_b.value = f"<b>Implied Growth B: {implied_growth_b*100:.2f}%</b>"

        # Calculate and display ROIC - COC spread
        spread_a_val = roica_val - coca_val
        spread_b_val = roicb_val - cocb_val
        color_a = "green" if spread_a_val >= 0 else "red"
        color_b = "green" if spread_b_val >= 0 else "red"

        roic_coc_spread_a.value = f"<b>ROIC - COC Spread A: <span style='color:{color_a};'>{spread_a_val:.2f}%</span></b>"
        roic_coc_spread_b.value = f"<b>ROIC - COC Spread B: <span style='color:{color_b};'>{spread_b_val:.2f}%</span></b>"


        df_projections = generate_projections(
            cost_of_capital_a=coca_val, roic_a=roica_val, inv_rate_a=inva_val,
            cost_of_capital_b=cocb_val, roic_b=roicb_val, inv_rate_b=invb_val,
        )

        with output_table:
            clear_output(wait=True)
            display(df_projections.round(2))

        with output_graphs:
            clear_output(wait=True)
            if df_projections is not None: # Ensure df is generated
                periods = df_projections.index[:-1].astype(str)
                pv_fcf_a = df_projections[('Firm A', 'PV of FCF')].iloc[:-1].values
                pv_fcf_b = df_projections[('Firm B', 'PV of FCF')].iloc[:-1].values
                cum_pv_fcf_a = np.cumsum(pv_fcf_a)
                cum_pv_fcf_b = np.cumsum(pv_fcf_b)

                fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(14, 6))

                ax1.plot(periods, pv_fcf_a, marker='o', linestyle='-', label='Firm A PV of FCF')
                ax1.plot(periods, pv_fcf_b, marker='x', linestyle='--', label='Firm B PV of FCF')
                ax1.set_xlabel("Period")
                ax1.set_ylabel("PV of FCF")
                ax1.set_title("PV of FCF per Period")
                ax1.legend()
                ax1.grid(True, linestyle='--', alpha=0.7)
                ax1.tick_params(axis='x', rotation=45)
                # Add horizontal line at y=0 for reference
                ax1.axhline(0, color='black', linewidth=0.5, linestyle='--')


                ax2.plot(periods, cum_pv_fcf_a, marker='o', linestyle='-', label='Firm A Cum. PV of FCF')
                ax2.plot(periods, cum_pv_fcf_b, marker='x', linestyle='--', label='Firm B Cum. PV of FCF')
                ax2.set_xlabel("Period")
                ax2.set_ylabel("Cumulative PV of FCF")
                ax2.set_title("Cumulative PV of FCF over Time")
                ax2.legend()
                ax2.grid(True, linestyle='--', alpha=0.7)
                ax2.tick_params(axis='x', rotation=45)
                ax2.axhline(0, color='black', linewidth=0.5, linestyle='--')


                plt.tight_layout()
                plt.show()
            else:
                print("Graphs cannot be generated as projections are missing.")


    except Exception as e:
        with output_table:
            clear_output(wait=True)
            print(f"Error processing input: {e}")
            print("Please ensure all inputs are valid numbers.")
        with output_graphs:
            clear_output(wait=True)
            print(f"Graphs cannot be generated due to input error: {e}")
        # Clear spread displays on error too
        growth_display_a.value = "Error"
        growth_display_b.value = "Error"
        roic_coc_spread_a.value = "Error"
        roic_coc_spread_b.value = "Error"

# --- Linking Button ---
def on_recalculate_button_clicked(b):
    update_outputs_and_graphs()

recalculate_button.on_click(on_recalculate_button_clicked)

# --- Display Layout ---
firm_a_label = widgets.HTML("<b>Firm A Inputs</b>")
box_a = widgets.VBox([firm_a_label, coc_a, roic_a, inv_rate_a, growth_display_a, roic_coc_spread_a])

firm_b_label = widgets.HTML("<b>Firm B Inputs</b>")
box_b = widgets.VBox([firm_b_label, coc_b, roic_b, inv_rate_b, growth_display_b, roic_coc_spread_b])

input_ui = widgets.HBox([box_a, box_b])

full_ui = widgets.VBox([
    input_ui,
    recalculate_button,
    output_explanation, # Added explanation output
    widgets.HTML("<hr><b>Projections Table:</b>"),
    output_table,
    widgets.HTML("<hr><b>Graphs:</b>"),
    output_graphs
])

display(full_ui)
update_outputs_and_graphs() # Initial call
```

### Investment Decision

*   If ROIC is larger than the cost of capital, higher reinvestment rate (and hence higher growth) increases value.
*   If the cost of capital is larger than ROIC, higher reinvestment rate or higher growth reduces value.
    *   This is because more investment in negative NPV project destroys value.
*   Once a firm has become "stable" or settled down in an equilibrium in which its reinvestment rate and ROIC are constant, its continuation value depends only on ROIC, cost of capital, and reinvestment rate.

### How To Increase Value
*   Increase ROIC by increasing operating margins or capital turnover.
*   Reduce the cost of capital.
*   Increase reinvestment rate as long as ROIC is larger than the cost of capital. Reduce it otherwise.

### Control of the Corporation and the Agency Problem

*   In large corporations, ownership can be spread over a huge number of stockholders.
    *   The dispersion of ownership arguably means that management effectively controls the firm.
    *   Does management necessarily act in the best interests of the stockholders?
    *   Why would not management pursue its own goals at the stockholders' expense?
*   The relationship between owner(s) and management is called an agency relationship.
*   Agency problem
    *   Exists whenever someone hires another to represent his or her interests.
    *   In all such relationships there is a possibility of a conflict of interest between the principal and the agent.
    *   If both parties to the relationship are utility maximizers there is good reason to believe that the agent will not always act in the best interests of the principal.
    *   Such a conflict is called an agency problem.
*   Agency costs arise whenever there is an incomplete alignment of the agent's and owner's interests.
    *   The monitoring expenditures by the principal,
    *   The bonding expenditures by the agent,
    *   The residual loss.

---

### Addressing Agency Problems
*   Whether the managers act in the stockholders' interests or not depends on two factors.
    *   How closely are management goals aligned with stockholder goals? (Managerial compensation)
    *   Can managers be replaced if they do not pursue stockholder goals? (Control of the firm)
*   Managerial Compensation
    *   Managerial compensation is tied to financial performance in general and often to share value.
    *   Managers who are successful in pursuing stockholder goals will have better job prospects since they will be in greater demand in the labor market and command higher salaries.
*   Control of the Firm
    *   Control of the firm ultimately rests with stockholders since they elect the board of directors, who, in turn, hire and fire management.
    *   An important mechanism by which unhappy stockholders can replace existing management is called a proxy fight.
        *   A proxy is the authority to vote someone else's stock.
        *   A proxy fight develops when a group solicits proxies in order to replace the existing board and thereby replace existing management.
    *   Takeovers.
        *   Firms that are poorly managed are more attractive as acquisitions than well-managed firms because a greater profit potential exists.
        *   Avoiding a takeover by another firm gives management another incentive to act in the stockholders' interests.
