### Corporate Governance: Cadbury Report

* Corporate governance is the system by which companies are directed and controlled.

* Boards of directors are responsible for the governance of their companies.

* The shareholders' role in governance is to appoint the directors and the auditors and to satisfy themselves that an appropriate governance structure is in place.

* Financial aspects of corporate governance are the way in which
    
    * boards set financial policy and oversee its implementation, including the use of financial controls,
    
    * and the process whereby they report on the activities and progress of the company to the shareholders.

---

### Corporate Governance
* Corporate governance refers to the
    * set/systems of responsibilities and practices exercised by the board and executive management

    * with the goal of providing strategic direction,

    * ensuring that objectives are achieved,

    * ascertaining that risks are managed appropriately, and

    * verifying that organization's resources are used responsibly.

* Systems of decision-making processes and of internal control, configuration of assets and related matters.

* Closely related to the of conflicts of interest and attempts to minimize them.

* What may be the examples of conflicts of interest?

* Effective corporate governance requires a clear understanding of
    * the respective roles of the board, management and shareholders;

    * their relationships with each other;

    * and their relationships with other corporate stakeholders.

### Corporate Governance: Key Corporate Actors
* **Management**
    * Responsible for setting, managing, and executing the strategies of the company, including but not limited to running the operations of the company under the oversight of the board and keeping the board informed of the status of the company's operations.
    
    * An effective management team runs the company with a focus on executing the company's strategy over a meaningful time horizon and avoids an undue emphasis on short-term metrics.

* **Management's responsibilities**
    * Business operations, strategic planning, capital allocation.
    
    * Identifying, evaluating, and managing risks
    
    * Accurate and transparent financial reporting and disclosures
    
    * Selecting qualified management, establishing an effective organizational structure, and ensuring effective succession planning

* **Shareholders**
    * Invest in a corporation by buying its stock and receive economic benefits in return.
    
    * Not involved in the day-to-day management of business operations
    
    * They have the right to elect the directors and to receive information material for investment and voting decisions.
    
    * Expect corporate boards and managers to act as long-term stewards of their investment in the corporation.
    
    * Expect that the board and management will be responsive to issues and concerns that are of widespread interest to long-term shareholders and affect the company's long-term value.

### Conflicts of Interest and the Agency Problem
* The separation of ownership and control is perhaps the most important reason for the success of the corporate organizational form.
    
    * The conflict of interest between managers and investors derives from the separation of ownership and control in a corporation.
    
    * The role of the corporate governance system is to mitigate the conflict of interest that results from the separation of ownership and control without unduly burdening managers with the risk of the firm. 
        * The system attempts to align these interests by providing incentives for taking the right action and punishments for taking the wrong action.

* **Agency problem**
    * The relationship between owner(s) and management is called an agency relationship.
        * Exists whenever someone hires another to represent his or her interests.
        * In all such relationships there is a possibility of a conflict of interest between the principal and the agent.
        * If both parties to the relationship are utility maximizers there is good reason to believe that the agent will not always act in the best interests of the principal.
        * Such a conflict is called an agency problem.
* **Agency costs arise whenever there is an incomplete alignment of the agent's and owner's interests.**
    * The monitoring expenditures by the principal,
    * The bonding expenditures by the agent,
    * The residual loss.

### Managing Agency Conflict
* **Direct Action by Shareholders**
    * **Shareholder Voice**
       
        * Any shareholder can submit a resolution that is put to a vote at the annual meeting.
       
        * A resolution could direct the board to take a specific action, such as discontinuing investing in a particular line of business or country or removing a poison pill.
       
        * "No” votes
    
    * **Shareholder Approval**
    
      * Shareholders must approve many major actions taken by the board.
    
        * Mergers & Acquisitions
    
        * Say on pay
    
    * **Proxy Contests**
        * Shareholders can hold a proxy contest and introduce a rival slate of directors for election to the board
    
        * Proxy fights can push management to take the needed action

### Managing Agency Conflict
* **Management Entrenchment**
    * Managers counter disciplinary forces by entrenching themselves, that is, by making themselves valuable to shareholders and costly to replace.

        * Antitakeover statutes,
    
        * Poison pills,
    
        * Staggered boards,
    
        * Restrictions on the ability of shareholders to call special meetings themselves.
    
    * As managers become more entrenched, they may tend to avoid long-term investments that stretch beyond their own tenures, and those with uncertain cash flows.
    
    * Effective corporate governance, in general, reduces managerial entrenchment.

### Agency Theory vs. Stakeholder Theory
* How are the social welfare and the objectives of the modern corporation linked?
    
    * Conventional wisdom in business argues that corporations should have a singular focus, shareholder wealth maximization.
    
        * **Assumption:** Shareholder wealth maximization yields social welfare improvements as social actors pursue their various interests via profit-generating market transactions.
    
        * Focusing on a single corporate objective is necessary for any kind of functional corporate decision-making.
    
        * Rational decision-making requires a single objective(?)
    
        * Difficulties in multi-objective decision-making.
    
    * Does shareholder wealth maximization also maximize social welfare?
    
    * Should the corporations be managed for multiple and sometimes conflicting stakeholders' interests?

### Agency Theory vs. Stakeholder Theory

* **Agency Theory**

    * One of the most widely used theories in management.

    * Examines the relationship between two parties, the principal (owner) and the agent (manager) from a behavioral and a structural perspective.

    * Suggests that given the chance, agents will behave in a self-interested manner, behavior that may conflict with the principal's interest.

    * The underlying assumption of agency theory is based on the economic model of man.

        * Assumes that individuals will seek to maximize their own utility.

        * An agent is hired to maximize the principal's utility, which is the firm value/shareholder wealth.

        * Assumes that agents will instead behave opportunistically because they too are self-serving.

        * The principal enacts mechanisms to minimize losses to their own utility.

    * An agency problem is created when the interest of the principal and agent are misaligned, and the principal lacks the information to accurately assess the behavior of the agent (information asymmetry).

    * Agency problems can take the form of moral hazard or adverse selection.

        * Moral hazard refers to the situation where the agent lacks effort in the scope of the employment relationship.

        * Adverse selection refers to the situation where the agent lacks the ability and skills to competently behave in the scope of the employment relationship.

---

### Agency Theory vs. Stakeholder Theory
* **Stakeholder Theory**
   
    * Set of propositions that suggest that management of companies have obligations to some group of stakeholders.
   
        * A stakeholder can be a voluntary or involuntary risk bearer.
   
        * Voluntary stakeholders bear some form of risk because of having invested some form of capital in the organization.
   
        * Involuntary stakeholders are placed at risk because of the firm's activities.
   
    * Companies are so large and their impact on society so pervasive that they should discharge accountability to many more sectors of society than solely their shareholders.
   
        * The stakeholder theory of corporate governance focuses on the effect of corporate activity on all identifiable stakeholders of the corporation.
   
        * Managing for stakeholders involves attention to more than simply maximizing shareholder wealth.
   
    * Main purposes of stakeholder theory,
   
        * To help boards of directors and management understand their stakeholder environments and manage more effectively within the nexus of relationships that exists for their companies.
   
        * To help directors and managers improve the value of the consequences of their actions and minimize the harm to stakeholders.
   
    * Stakeholder theory provides a convincing theoretical framework for analyzing the relationship between a company and society and hence, is closely related to the literature on sustainability and corporate social responsibility, which we will cover in the last lecture of the course.

### The Objective Function
* In traditional corporate finance, the objective in decision-making is to maximize the value of the firm or maximize stockholder wealth.

* Why traditional corporate financial theory focuses on maximizing stockholder wealth?
   
    * You can have only one objective, i.e., one interest group, whose interests get placed first.
   
    * Corporate finance picks shareholders because they have a residual claim, whereas every other claim-holder has a contractual claim that they can negotiate to protect their interests.
    
    * The strength of the stockholder wealth maximization as the objective function is its internal self-correction mechanism.
   
    * Excesses generally lead to counteractions that reduce or eliminate these excesses.
        
        * Managers taking advantage of stockholders has led to a much more active market for corporate control.
        
        * Stockholders taking advantage of bondholders has led to bondholders protecting themselves at the time of the issue.
        
        * Firms revealing incorrect or delayed information to markets has led to markets becoming more “skeptical” and "punitive"
        
        * Firms creating social costs have led to more regulations, as well as investor and customer backlashes.

---

### The Objective Function
* An alternative to the traditional objective is maximizing the collective wealth of all stakeholders.
    
    * Has been attaining more acceptance with the increased focus on sustainability and social responsibility.
    
    * The traditional finance theory is built upon utility maximization by economic agents, which is proxied by wealth maximization.
    
    * We need to find a new proxy that reflects the collective stakeholder wealth.
    
    * A more proper way is appending utility related to the social welfare to the utility function of economic agents, which is a daunting task and requires upheaval of the finance theory.
    
    * In practice, a common way of dealing with the issue is maximizing stockholder wealth subject to constraints related to social welfare, which may not be efficient.

* What are the traditional approach's counterarguments?

    * Maximizing stock price is not incompatible with meeting employee needs/objectives.

        * Employees are often stockholders in many firms

        * Firms that maximize stock price generally are profitable firms that can afford to treat employees well.

    * Maximizing stockholder wealth does not mean that customers are not critical to success. For most businesses, keeping customers happy is the route to stock price maximization.

    * Maximizing stock price does not imply that a company must be a social outlaw.

### Value Maximization, Stakeholder Theory, and the Corporate Objective Function
* If we could dictate the criterion or objective function to be maximized by firms (and thus the performance criterion by which corporate executives choose among alternative policy options), what would it be?
   
    * Managers must have a criterion for evaluating performance and deciding between alternative courses of action, and the criterion should be the maximization of the long-term market value of the firm.
   
    * Stakeholder theory says that managers should make decisions that take into account the interests of all the stakeholders in a firm.
   
        * Stakeholders include all individuals or groups who can substantially affect, or be affected by, the welfare of the firm.

* What is the main argument against the stakeholder theory as a legitimate contender to value maximization?

    * It fails to provide a complete specification of the corporate purpose or objective function.

    * Stakeholder theory directs corporate managers to serve “many masters” whereas value maximization provides corporate managers with a single objective.

    * Without the clarity of mission provided by a single-valued objective function, companies embracing stakeholder theory will experience managerial confusion, conflict, inefficiency, and perhaps even competitive failure.

### Value Maximization, Stakeholder Theory, and the Corporate Objective Function
* Why does purposeful behavior require the existence of a single-valued objective function?

    * It is logically impossible to maximize in more than one dimension at the same time.

    * Telling a manager to maximize multi objectives will leave that manager with no way to make a reasoned decision, with no objective.

* Does the maximization of total firm value make society better off?

    * Social value is created whenever a firm produces an output, or set of outputs, that is valued by its customers at more than the value of the inputs it consumes in the production of the outputs.

    * Firm value is simply the long-term market value of this expected stream of benefits.

    * Social welfare is maximized when all firms in an economy attempt to maximize their own total firm value.

* When does the value-maximizing criterion not maximize social welfare?

    * Monopolies

    * Externalities

* How does the profit maximization increase social welfare?

    * When a company acquires an additional unit of any input(s) to produce an additional unit of any output, it increases social welfare by at least the amount of its profit.

    * Continue to expand purchases of inputs and sell the resulting outputs as long as an additional dollar of inputs generates sales of at least a dollar.

* How do the interest rates help us with the tradeoffs through time?

* How can managers deal with uncertainty and make decisions on tradeoffs?

* What is the main problem with the Stakeholder Theory's assertion that firms should pay attention to all their constituencies?

    * Customers want low prices, high quality, and full service.

    * Employees want high wages, high-quality working conditions, and fringe benefits, including vacations, medical benefits, and pensions.

    * Suppliers of capital want low risk and high returns.

    * Communities want high charitable contributions, social expenditures by companies to benefit the community at large, increased local investment, and stable employment.

    * Any theory of corporate decision-making must tell the decision-makers how to choose among multiple constituencies with competing and, in some cases, conflicting interests.

    * Any decision criterion must specify how to make the tradeoffs between these demands.

### Value Maximization, Stakeholder Theory, and the Corporate Objective Function
* **Enlightened Value Maximization**

    * Value maximization is not a vision or a strategy or even a purpose; it is the scorecard for the organization.

    * Value maximization or value seeking tells an organization and its participants how their success in achieving a vision or in implementing a strategy will be assessed.

    * We cannot maximize the long-term market value of an organization if we ignore or mistreat any important constituency.

    * Stakeholder Theory may be helpful in leading the managers and participants in an organization to think more generally and creatively about how the organization's policies treat all important constituencies of the firm.
