*   In practice, dividend policy is dysfunctional and does not follow the logical process of starting with your investment opportunities and working your way down to residual cash.

*   The two dominant factors driving dividend policy around the world are:
    *   Inertia: Companies seem to hate to let go of their past, when it comes to dividend policy.
    *   Me-too-ism: Companies want to behave like their peer group.

### Dividends
*   **Declaration Date**
    *   The date on which the board of directors authorizes the payment of a dividend.

*   **Record Date**
    *   When a firm pays a dividend, only shareholders on record on this date receive the dividend.

*   **Ex-dividend Date**
    *   A date, two business days prior to a dividend's record date, on or after which anyone buying the stock will not be eligible for the dividend.

*   **Payable Date (Distribution Date)**
    *   A date, generally within a month after the record date, on which a firm mails dividend checks to its registered stockholders.

---

### Dividends

*   **Special Dividend**
    *   A one-time dividend payment a firm makes, which is usually much larger than a regular dividend

*   **Stock Dividend**
    *   When a company issues a dividend in shares of stock rather than cash to its shareholders

*   **Return of Capital**
    *   When a firm, instead of paying dividends out of current earnings (or accumulated retained earnings), pays dividends from other sources, such as paid-in-capital or the liquidation of assets

*   **Liquidating Dividend**
    *   A return of capital to shareholders from a business operation that is being terminated

### Share Repurchases

*   An alternative way to pay cash to investors is through a share repurchase or buyback.

*   **Open Market Repurchase**
    *   When a firm repurchases shares by buying shares in the open market
    *   Open market share repurchases represent about 95% of all repurchase transactions.

*   **Tender Offer**
    *   A public announcement of an offer to all existing security holders to buy back a specified number of outstanding securities at a prespecified price over a prespecified period of time.
    *   If shareholders do not tender enough shares, the firm may cancel the offer, and no buyback occurs.

*   **Dutch Auction**
    *   A share repurchase method in which the firm lists different prices at which it is prepared to buy shares, and shareholders in turn indicate how many shares they are willing to sell at each price.
    *   The firm then pays the lowest price at which it can buy back its desired number of shares.

*   **Targeted Repurchase**
    *   When a firm purchases shares directly from a specific shareholder

*   **Greenmail**
    *   When a firm avoids a threat of takeover and removal of its management by a major shareholder by buying out the shareholder, often at a large premium over the current market price.

### Dividends vs. Share Repurchase

Lego's board is deciding how to distribute €30 million in excess cash to its shareholders. Lego currently has no debt, and its equity cost of capital (which equals its unlevered cost of capital) is 10%. The company has 15 million shares outstanding. After this immediate distribution, Lego expects to generate future free cash flows of €75 million per year indefinitely, and it plans to pay out all these future FCFs to shareholders.

1.  If Lego decides to pay out the €30 million as a special dividend today:
    a.  What is the dividend per share ($D_0$)?
    b.  What will be Lego's cum-dividend share price ($P_{cum}$)?
    c.  What will be Lego's ex-dividend share price ($P_{ex}$)?

2.  Suppose instead, Lego uses the €30 million to repurchase its shares on the open market.
    a.  If the shares are repurchased at the fair cum-dividend price calculated in 1.b, how many shares will be repurchased?
    b.  How many shares will remain outstanding after the repurchase?
    c.  What will be the expected future cash flow per share available to the remaining shareholders?
    d.  What will be Lego's share price after the repurchase ($P_{rep}$)?

### Dividends vs. Share Repurchase

**Special Dividend Payout**

$D_0 = \frac{\text{Total Special Dividend}}{\text{Shares Outstanding}} = \frac{€30,000,000}{15,000,000} = €2.00 $

**Cum-dividend share price ($P_{cum}$):**
    
$PV(\text{Future Payouts per Share}) = \frac{\text{Total Future FCF per year} / N_0}{r_E} = \frac{€75,000,000 / 15,000,000}{0.10} = \frac{€5.00}{0.10} = €50.00$
        
$ P_{cum} = D_0 + PV(\text{Future Payouts per Share}) \quad \Rightarrow \quad P_{cum} = €2.00 + €50.00 = €52.00 $

**Ex-dividend share price:** $\quad P_{ex} = PV(\text{Future Payouts per Share}) = €50.00 $


**Share Repurchase**

Shares repurchased (assuming repurchase at $P_{cum} = €52.00$):

$ \text{Shares Repurchased} = \frac{\text{Cash for Repurchase}}{\text{Repurchase Price}} = \frac{€30,000,000}{€52.00/\text{share}} \approx 576,923.08 \text{ shares} $

Shares remaining outstanding:

$ N_1 = N_0 - \text{Shares Repurchased} = 15,000,000 - 576,923.08 \approx 14,423,076.92 \text{ shares} $

Expected future cash flow per share for remaining shareholders:

$CFPS_1 = \frac{€75M}{750M/52} = \frac{€75M \times 52}{750M} = \frac{€52}{10} = €5.20$ 

Share price after repurchase:

$ P_{rep} = \frac{CFPS_1}{r_E} = \frac{€5.20}{0.10} = €52.00 $

### Payout Policies

*   In perfect capital markets, 
    * An open market share repurchase has no effect on the stock price, and the stock price is the same as the cum-dividend price if a dividend were paid instead.
    
    * Investors are indifferent between the firm distributing funds via dividends or share repurchases.
      *   By reinvesting dividends or selling shares, they can replicate either payout method on their own.

*   The relationship between the price change on the ex-dividend day and the dollar dividend will be determined by the difference between the tax rate on dividends and the tax rate on capital gains for the typical investor in the stock.
    * The price drop can be expressed as: $\Delta P = D \times \frac{1 - \tau_D}{1 - \tau_{CG}}$
    where $\tau_D$ is the tax rate on dividends and $\tau_{CG}$ is the tax rate on capital gains.

| Tax Rates Scenario                                           | Ex-Dividend Day Price Drop ($\Delta P$) Behavior | Explanation (Investor Perspective)                                                                                                |
| :----------------------------------------------------------- | :---------------------------------------------: | :-------------------------------------------------------------------------------------------------------------------------------- |
| Dividends and capital gains taxed equally ($\tau_D = \tau_{CG}$) |                   $\Delta P = D$                  | The after-tax value of the dividend equals the after-tax value of the capital gain avoided by selling before the price drop. Price drops by the full dividend. |
| Dividends taxed higher than capital gains ($\tau_D > \tau_{CG}$) |                   $\Delta P < D$                  | To be indifferent, the price must drop by less than the dividend, as dividends are less attractive post-tax. Selling cum-dividend is tax-preferred. |
| Dividends taxed lower than capital gains ($\tau_D < \tau_{CG}$)  |                   $\Delta P > D$                  | To be indifferent, the price must drop by more than the dividend, as dividends are more attractive post-tax. Receiving the dividend is tax-preferred.  |

### Payout Policies

*   Dividends do not matter, and dividend policy does not affect value.
    *   If there are no tax disadvantages associated with dividends & companies can issue stock, at no issuance cost, to raise equity, whenever needed

*   Dividends are bad, and increasing dividends will reduce value
    *   If dividends create a tax disadvantage for investors (relative to capital gains)

*   Dividends are good, and increasing dividends will increase value
    *   If dividends create a tax advantage for investors (relative to capital gains) and/or stockholders like dividends

*   If a company has excess cash, and no good investment opportunities (NPV>0), returning money to stockholders is good.

*   If a company does not have excess cash, and/or has several good investment opportunities (NPV>0), returning money to stockholders is bad.

### Payout vs. Retention of Cash

*   In perfect capital markets, once a firm has taken all positive-NPV investments, it is indifferent between saving excess cash and paying it out.
    
    *   If a firm has already taken all positive-NPV projects, any additional projects it takes on are zero or negative-NPV investments.
    
    *   Rather than waste excess cash on negative-NPV projects, a firm can use the cash to purchase financial assets.
    
    *   In perfect capital markets, buying and selling securities is a zero-NPV transaction, so it should not affect firm value.

*   With market imperfections, there is a tradeoff: 
    
    * Retaining cash can reduce the costs of raising capital in the future, but it can also increase taxes and agency costs.

*   Aloha Mining has $100,000 in excess cash and considering investing the cash in one-year Treasury bills paying 6% interest, and then using the cash to pay a dividend next year.
    
    *   Alternatively, the firm can pay a dividend immediately and shareholders can invest the cash on their own.
    
    *   In a perfect capital market, which option will shareholders prefer?

### Payout vs. Retention of Cash

|                               | Quality of projects taken: Poor projects                                             | Quality of projects taken: Good projects                               |
| :---------------------------------------------- | :---------------------------------------------------------------------- | :------------------------------------------------------------------------------- |
| **Cash Returned < FCFE** (Cash Surplus)                        | **Cash Surplus + Poor Projects**<br/>Significant pressure to pay out more to stockholders as dividends or stock buybacks | **Cash Surplus + Good Projects**<br/>Maximum flexibility in setting dividend policy |
| **Cash returned > FCFE** (Cash Deficit relative to Free Cash flow to Equity) | **Cash Deficit + Poor Projects**<br/>Reduce or eliminate cash return but real problem is in investment policy. | **Cash Deficit + Good Projects**<br/>Reduce cash payout, if any, to stockholders |

*   Microsoft had accumulated a cash balance of $ 43 billion by 2002 by paying out no dividends while generating huge FCFE.
    *   At the end of 2003, there was no evidence that Microsoft was being penalized for holding such a large cash balance or that stockholders were becoming restive about the cash balance. There was no demand for more dividends or stock buybacks. Why?
    *   In 2004, Microsoft announced a huge special dividend of $ 32 billion and made clear that it would try to return more cash to stockholders in the future. What do you think changed?
    *   Microsoft's special dividend of $3 per share, or $32 billion, during late 2004. If Microsoft had instead retained that cash permanently, what would the present value of the additional taxes be?
    $$\frac{\$32 \text{ billion} \times r_f \times 35\%}{r_f} = \$32 \text{ billion} \times 35\% = \$11.2 \text{ billion}$$

### Payout Policies

*   Generally, firms retain cash balances to cover potential future cash shortfalls, despite the tax disadvantage to retaining cash.
    
    *   A firm might accumulate a large cash balance if there is a reasonable chance that future earnings will be insufficient to fund future positive-NPV investment opportunities.
    
    *   The cost of holding cash to cover future potential cash needs should be compared to the reduction in transaction, agency, and adverse selection costs of raising new capital through new debt or equity issues.

*   **Agency Costs of Retaining Cash**
    
    *   When firms have excessive cash, managers may use the funds inefficiently by paying excessive executive perks, over-paying for acquisitions, etc.
    
    *   Paying out excess cash through dividends or share repurchases, rather than retaining cash, can boost the stock price by reducing managers' ability and temptation to waste resources.
    
    *   Firms should choose to retain to help with future growth opportunities and to avoid financial distress costs.

### Payout Policies

*   **The Miller-Modigliani Hypothesis: Dividends do not affect value**
    
    *   If a firm's investment policies (and hence cash flows) don't change, the value of the firm cannot change as it changes dividends.
    
    *   If a firm pays more in dividends, it will have to issue new equity to fund the same projects. By doing so, it will reduce expected price appreciation on the stock, but it will be offset by a higher dividend yield.
    
    *   If we ignore personal taxes, investors have to be indifferent to receiving either dividends or capital gains.

    *   **Underlying Assumptions:**
        
        *   There are no tax differences to investors between dividends and capital gains.
        
        *   If companies pay too much in cash, they can issue new stock, with no flotation costs or signaling consequences, to replace this cash.
        
        *   If companies pay too little in dividends, they do not use the excess cash for bad projects or acquisitions.

### Payout Policies

*   **The bird in the hand fallacy**
    
    *   Dividends now are more certain than capital gains later. Hence dividends are more valuable than capital gains. Stocks that pay dividends will therefore be more highly valued than stocks that do not.
    
    *   The appropriate comparison should be between dividends today and price appreciation today. The stock price drops on the ex-dividend day.

*   **Excess cash**
    
    *   The firm has excess cash on its hands this year, no investment projects this year and wants to give the money back to stockholders.
    
    *   Why not just repurchase stock?
        
        *   If this is a one-time phenomenon, the firm has to consider future financing needs.
        
        *   The cost of raising new financing in future years, especially by issuing new equity, can be staggering.

*   **The Signalling Story:** Dividends can be signals to the market that you believe that you have good cash flow prospects in the future.

*   **The Wealth Appropriation Story:** Dividends are one way of transferring wealth from lenders to equity investors (this is good for equity investors but bad for lenders)

*   **Clientele Effect:** The investors in your company like dividends.
    
    *   Investors may form clienteles based upon their tax brackets. Investors in high tax brackets may invest in stocks which do not pay dividends and those in low tax brackets may invest in dividend paying stocks.

### Payout Policies: Clientele Effect

*   When the dividend policy of a firm reflects the tax preference of its investor clientele

*   Individuals in the highest tax brackets have a preference for stocks that pay no or low dividends, whereas tax-free investors and corporations have a preference for stocks with high dividends.

| Investor Group              | Dividend Policy Preference                                                               | Proportion of Investors |
| :-------------------------- | :--------------------------------------------------------------------------------------- | ----------------------: |
| Individual investors        | Tax disadvantage for dividends generally prefer share repurchase (except for retirement accounts) | ~52%                    |
| Institutions, pension funds | No tax preference prefer dividend policy that matches income needs                       | ~47%                    |
| Corporations                | Tax advantage for dividends                                                              | ~1%                     |

*   High growth firms are sometimes advised to initiate dividends because its increases the potential stockholder base for the company and, by extension, the stock price.

### Payout Policies: Clientele Effect

*   **NVIDIA (NVDA):**
    *   NVIDIA is a prime example of a company with massive R&D needs and enormous growth prospects, historically reinvesting nearly all profits.
    *   Yet, NVIDIA has maintained a policy of paying a cash dividend, albeit a very small one relative to its earnings and stock price (resulting in a tiny dividend yield).

*   **Why would a high-growth company like NVIDIA choose to pay any dividend at all, even if nominal?**
    
    *   **Signaling Hypothesis:** 
        * Does it signal confidence in sustained profitability and cash flow generation, even amidst heavy reinvestment?
        
    *   **Catering to a Broader Investor Base (Clientele):** 
        * Does even a small dividend make the stock eligible for certain investment funds, indices, or "dividend-paying stock" lists, thereby expanding potential ownership?
        
    *   **"Widow and Orphan Stock" Perception (Legacy/Mature Growth):** 
        * Could it be a nod towards appearing as a stable, "blue-chip" type company, even if its primary characteristic is growth?
        
    *   **Discipline on Management:** 
        * Does it impose a discipline on cash management?
        
    *   **Behavioral Factors:** 
        * Do some retail investors simply prefer to see some cash return, regardless of size?
    
*   Does the existence of such a nominal dividend policy challenge or refine our understanding of the "optimal" payout policy for purely growth-oriented firms?

### Dividend-Capture Theory
*   The theory states that when transaction cost are absent, investors can trade shares at the time of the dividend so that non-taxed investors receive the dividend.

*   An implication of this theory is that we should see large trading volume in a stock around the ex-dividend day, as high-tax investors sell and low-tax investors buy the stock in anticipation of the dividend, and then reverse those trades just after the ex-dividend date.

### The Effective Dividend Tax Rate
*   Consider buying a stock just before it goes ex-dividend and selling the stock just after. The equilibrium condition must be
    $$(P_{cum} - P_{ex}) (1-\tau_g) = \text{Div}(1 - \tau_d)$$
    $$P_{cum} - P_{ex} = \text{Div} \times \left( \frac{1-\tau_d}{1-\tau_g} \right) = \text{Div} \times \left( 1 - \frac{\tau_d - \tau_g}{1-\tau_g} \right) = \text{Div} \times (1 - \tau_d^*)$$
    Where the effective dividend tax rate $\tau_d^*$ is:
    $$\tau_d^* = \frac{\tau_d - \tau_g}{1-\tau_g}$$

### Payout Policies

*   Assume a firm raises $25 million from shareholders and uses this cash to pay them $25 million in dividends.
    *   Dividends are taxed at a 39% tax rate.
    *   Capital gains are taxed at a 20% tax rate.

*   How much will shareholders receive after taxes?
    *   On dividends, shareholders will owe
        *   39% × $25 million = $9.75 million in dividend taxes.
    
    *   Shareholders will lower their capital gains taxes by
        *   20% × $25 million = $5 million
        *   The value of the firm will fall when the dividend is paid, lowering the shareholders' capital gains.
    
    *   Shareholders will pay a total of $4.75 million in taxes.
        *   $9.75 - $5.00 = $4.75 million
    
    *   Shareholders will receive back only $20.25 million of their $25 million investment.
        *   $25 - $4.75 = $20.25 million

*   When the tax rate on dividends is greater than the tax rate on capital gains, shareholders will pay lower taxes if a firm uses share repurchases rather than dividends.
    *   Tax savings will increase the value of a firm that uses share repurchases rather than dividends.
*   The optimal dividend policy when the dividend tax rate exceeds the capital gain tax rate is to pay no dividends at all.

### Case Study: Omega Digital - The Payout Predicament

**Case Date:** November 2024

Omega Digital, a moderately mature software company, has recently reported another year of solid profitability. The company has a history of reinvesting most of its earnings into R&D and strategic acquisitions, which has fueled its growth over the past decade. However, its super-normal growth phase appears to be moderating. Omega currently has no debt and a significant cash balance of $75 million on its balance sheet, which analysts deem "excess cash" beyond its immediate operational and strategic investment needs.

The company has 10 million shares outstanding, and its stock has performed well, recently trading around $60 per share. Omega has never paid a regular cash dividend, occasionally using share repurchases to return capital, though not on a consistent schedule.

**The Board Meeting:**

CEO Eva Rostova has called a special board meeting to discuss Omega's capital return policy. The agenda is clear: **"What should Omega Digital do with its growing excess cash pile, and what should our ongoing payout policy look like?"**

During the meeting, several viewpoints emerge:

*   **Director Chen:** "We've built a successful company. Our earnings are stable. It's time we started paying a regular quarterly cash dividend. It signals maturity, financial health, and it will attract a new class of investors who look for income. Our shareholders deserve a regular return."

*   **Director Miller:** "Hold on. Dividends are often tax-inefficient for many of our individual shareholders compared to capital gains. If we have excess cash, shouldn't we prioritize share repurchases? It gives shareholders flexibility and likely a better after-tax return. Why lock ourselves into a rigid dividend schedule?"

*   **Director Alvi:** "Are we sure all high-return investment opportunities are exhausted? That $75 million could fund a significant R&D push into next-gen AI integration for our software, or a strategic acquisition. Paying it out feels like we're signaling our growth story is over. We are a tech company, after all."

*   **CFO Ben Carter:** "From a financial flexibility standpoint, holding some excess cash has served us well, especially given the dynamic nature of the tech industry. However, $75 million is substantial. If we do initiate a regular dividend, we need to be confident we can maintain or grow it – cutting a dividend is seen very negatively. Repurchases offer more flexibility. Regarding new investments, our current pipeline of known high-NPV projects doesn't require the entirety of this cash."

**Market Context & Investor Profile:**

*   Omega's shareholder base is diverse:
    *   Significant portion held by institutional investors (some tax-exempt pension funds, others growth-focused mutual funds).
    *   A growing number of retail investors, many in higher tax brackets.
    *   A small percentage held by corporate investors.
*   Current Tax Environment:
    *   Qualified Dividends Tax Rate (for individuals): 15-20% (depending on income)
    *   Long-Term Capital Gains Tax Rate (for individuals): 15-20%
    *   Corporate Tax Rate (for Omega): 25%
    *   Assume for simplicity that for most individual investors, the tax rate on qualified dividends and long-term capital gains is currently similar.

**Discussion Points for Class:**

1.  **MM Irrelevance Proposition:** In a "perfect world", would Omega's choice between dividends, share repurchases, or retaining cash to invest in financial assets with NPV=0 affect its total firm value or share price? Explain the core logic.

2.  **The Case for Dividends (Director Chen):**
    a.  What are the potential benefits of Omega initiating a regular cash dividend, considering its current stage and cash position?
    b.  What potential downsides or risks should Omega consider if it commits to a regular dividend?

3.  **The Case for Share Repurchases (Director Miller):**
    a.  Why might share repurchases be considered more tax-efficient than dividends, even if current tax rates on qualified dividends and capital gains are similar for individuals?
    b.  Beyond taxes, what other advantages do share repurchases offer the company and its shareholders compared to dividends?

4.  **The Case for Reinvestment/Cash Retention (Director Alvi & CFO Carter):**
    a.  Under what conditions would retaining the $75 million cash or a significant portion of it be the best strategy for maximizing shareholder value?
    b.  What are the potential agency costs associated with Omega holding a large excess cash balance?

5.  **Clientele Effect & Signaling:**
    a.  How might Omega's current shareholder base its "clientele" react to a shift towards regular dividends versus a more aggressive share repurchase program?
    b.  What "signal" might initiating a regular dividend send to the market about Omega's future prospects? What signal might a large, opportunistic share repurchase send?

6.  **Decision:**
    a.  If you were CEO Eva Rostova, what key questions would you ask your team to help guide the decision-making process?
    b.  Considering Omega's situation (maturing tech, excess cash, diverse investors), what might be a balanced and prudent payout strategy for the company to consider for the $75 million and for its ongoing policy? Justify your recommendation.
