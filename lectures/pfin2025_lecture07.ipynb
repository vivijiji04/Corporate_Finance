### The Discounted Free Cash Flow Model - FCFE
*   The FCFE is the residual cash flow left over after meeting interest and principal payments and providing for reinvestment to maintain existing assets and create new assets for future growth.

*   The equity cash flow model values equity directly by discounting cash flows to equity (CFE) at the cost of equity, rather than at the weighted average cost of capital.
    $$ FCFE = \text{Net Income} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} – \text{Principal Payments} + \text{New Debt Issues} $$
    $$ V_0 = \frac{FCFE_1}{(1+r_e)^1} + \frac{FCFE_2}{(1+r_e)^2} + \dots + \frac{FCFE_N + V_N}{(1+r_e)^N} \quad \text{where } V_0 \text{ is the value of equity} $$
    $$ P_0 = \frac{V_0}{\text{Shares Outstanding}_0} $$

*   When the capital expenditures and the working capital are financed at a fixed (target) debt ratio $\delta$ and principal repayments are made from new debt issues:
    $$ FCFE = \text{Net Income} – (1 – \delta)(\text{CapEx} – \text{Depreciation}) – (1 – \delta)\Delta \text{NWC} $$

*   If we define the portion of the net income that equity investors reinvest back into the firm as the equity reinvestment rate, we can state the FCFE as a function of this rate.
    $$ \text{Equity Reinvestment Rate} = \frac{(\text{CapEx} – \text{Depreciation} + \Delta \text{NWC})(1 – \delta)}{\text{Net Income}} $$
    $$ FCFE = \text{Net Income}(1 – \text{Equity Reinvestment Rate}) $$

### The Discounted Free Cash Flow Model - FCFE

Free Cash Flow to Equity (FCFE) can be derived from Free Cash Flow to Firm (FCFF) by accounting for cash flows to/from debt holders and the tax shield on interest:

$$ FCFE = FCFF - \text{Interest Expense} \times (1 - \tau_C) + \text{Net Borrowing} $$


**Reconciliation Using Our Scenarios (Year 1 Data):**

**Common Data (Year 1):**
*   FCFF (Year 1) = $118.75 million
*   Interest Expense (Year 1) = $62.40 million
*   Corporate Tax Rate ($\tau_C$) = 35%
*   After-Tax Interest Expense = $62.40 \times (1 - 0.35) = \$62.40 \times 0.65 = \$40.56 \text{ million}$

**Scenario 1: Zero Net Borrowing**

$$FCFE_1 = FCFF_1 - \text{Interest}_1(1-\tau_C) + \text{Net Borrowing}_1 \quad \Rightarrow \quad FCFE_1 = \$118.75 - \$40.56 + \$0 \quad \Rightarrow \quad FCFE_1 = \$78.19 \text{ million}$$

**Scenario 2: Target D/V = 40% for New Investments (from Table 2)**

*   Net Investment (Year 1) = CAPEX - Depr + $\Delta$ NWC  
    $= \$133.75 - \$100.00 + \$26.25 = \$60.00 \text{ million}$
*   Assumed Net Borrowing (Year 1) = Net Investment $\times$ Target D/V Ratio  
    $= \$60.00 \text{ million} \times 0.40 = \$24.00 \text{ million}$

$$FCFE_1 = FCFF_1 - \text{Interest}_1(1-\tau_C) + \text{Net Borrowing}_1 \quad \Rightarrow \quad FCFE_1 = \$118.75 - \$40.56 + \$24.00 \quad \Rightarrow \quad FCFE_1 = \$102.19 \text{ million}$$

### The Discounted Free Cash Flow Model - FCFE

1.  **Net Borrowing Assumption:** This is the **primary driver** of the difference in FCFE between the two scenarios.  
    *   Scenario 1 (Table 1) assumes the firm issues no new net debt (Net Borrowing = 0). All investments beyond what's covered by non-cash charges (like depreciation) effectively reduce FCFE or require new equity.  
    *   Scenario 2 (Table 2) assumes the firm issues new debt to finance a specific portion (40%) of its net investments, thus maintaining a target leverage ratio for *new* financing. This increases the cash flow available to equity holders (FCFE) compared to Scenario 1.

2.  **Impact on Equity Value ($V_0$):**  
    *   Because Scenario 2 has higher FCFEs (due to positive net borrowing), its resulting Equity Value ($V_0 \approx \$1{,}643 \text{ million}$) is higher than the equity value for Scenario 1 ($V_0 \approx \$1{,}517 \text{ million}$), assuming the same cost of equity ($r_E = 13.25\%$) is used.

3.  **Consistency of Cost of Equity ($r_E$):**  
    *   The assumed cost of equity ($r_E$) should be consistent with the leverage implied by the financing policy.  
        *   In Scenario 1 (Net Borrowing = 0), if the firm has existing debt and equity grows via retained earnings, the D/E ratio will likely decrease over time. This should, theoretically, lead to a decreasing $r_E$. Using a constant $r_E = 13.25\%$ is a simplification.  
        *   In Scenario 2 (Net Borrowing to maintain D/V = 40% for new investments), if this policy leads to the overall firm D/V ratio converging to and stabilizing at 40%, then an $r_E$ appropriate for a 40% D/V firm is more consistent.  
        *   Using the same $r_E$ for both scenarios when their leverage profiles diverge is a simplification that contributes to different valuation outcomes.

**FCFF vs. FCFE valuation**  
*   Should theoretically yield the same equity value if all assumptions (especially regarding financing, net borrowing, and the discount rates used) are internally consistent and reflect the same underlying view of the firm's operations and financial policy.

### The Discounted Free Cash Flow Model - FCFE

*   Steps for computing FCFE
    *   Cash flow to equity starts with net income.
    *   Add back noncash expenses, and subtract investments in working capital, fixed assets, and nonoperating assets.
    *   Add any increases in debt and other nonequity claims and subtract decreases in debt and other nonequity claims.

*   Alternatively, you can compute cash flow to equity as dividends plus share repurchases minus new equity issues.

*   Although the dividends can never be less than zero, the FCFE can be negative. This can occur even if earnings are positive, if the firm has substantial working capital and capital expenditure needs.

*   Shortcomings of FCFE Model
    *   Easy to change the company's capital structure without realizing it when using the FCFE model.
        *   Increase debt to raise the dividend payout ratio, increased dividends lead to higher equity cash flows and a higher valuation.
    *   It values nonoperating assets.
        *   A company that holds a significant amount of low-risk, low-return excess cash.
    *   Requires allocating debt and interest expense to each unit when valuing a company by business unit.

### The Discounted Free Cash Flow Model - FCFE
*   FCFE and Dividend Discount Model
    *   The value obtained from using the FCFE in discounted cash flow valuation will be the same as the value obtained from using the dividend discount model.
        *   When the dividends are equal to the FCFE.
        *   When the FCFE is greater than dividends, but the excess cash (FCFE – Dividends) is invested in zero-NPV projects.
    
    *   When the FCFE is greater than the dividend and the excess cash either earns below-market returns or is invested in negative NPV projects, the value from the FCFE model will be greater than the value from the dividend discount model.
    
    *   When dividends are greater than FCFE, the firm will have to issue new shares or borrow money to pay these dividends. Then,
        *   The flotation cost on these security issues.
        *   If the firm borrows the money to pay the dividends, the firm may become overleveraged (relative to the optimal), leading to a loss in value.
        *   Paying too much in dividends can lead to capital rationing constraints, whereby good projects are rejected, resulting in a loss of wealth.

**Free Cash Flow to Firm (FCFF):**
$$ FCFF = EBIT \times (1 - \tau_c) + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} $$
    
Where: EBIT = \$605 million, $\tau_c = 40\%$, Depreciation = \$180 million, CapEx = \$349 million
    
$\Delta \text{NWC} = (\Delta \text{AR} + \Delta \text{Inv}) - (\Delta \text{AP} + \Delta \text{Accrued Exp.}) \quad \Rightarrow \quad \Delta \text{NWC} = (39 + 44) - (22 + 23) = 83 - 45 = 38 \text{ million}$

$ FCFF = 605 \times (1 - 0.40) + 180 - 349 - 38 \quad \Rightarrow \quad FCFF = 156 \text{ million} $

**Free Cash Flow to Equity (FCFE):**

*   Net Income = $285 million
*   Net Debt Issued = (New Debt Issues) - (Principal Repayments)
    *   Repayment of Notes Payable = $10 million
    *   Net Repayment of Long-Term Debt = $40 million
    *   Net Debt Issued = $0 - ($10 + $40) = -$50 million

$ FCFE = \text{NI} + \text{Depreciation} – \text{CapEx} – \Delta \text{NWC} + \text{Net Debt Issued} \\ \Rightarrow \quad FCFE = 285 + 180 - 349 - 38 + (-50) \quad \Rightarrow \quad FCFE = 28 \text{ million} $

**Adjustments from FCFF to FCFE:**

$ FCFE = FCFF - \text{Interest Expense} \times (1 - \tau_c) + \text{Net Debt Issued} $

*   FCFF = $156 million, Interest Expense = $130 million, $\tau_c = 0.40$, Net Debt Issued = -$50 million.

$$ FCFE = 156 - 130 \times (1 - 0.40) + (-50) \quad \Rightarrow \quad FCFE = 28 \text{ million} $$

### The Adjusted Present Value Approach
*   All discounted-cash-flow methodologies involve forecasting future cash flows and then discounting them to their present value at a rate that reflects their riskiness.

*   The methodologies differ in the details of their execution,
    *   Particularly, how the value created or destroyed by financial maneuvers is accounted for.
    *   APV's approach is to analyze financial maneuvers separately and then add their value to that of the business.

*   Like WACC, APV is designed to value operations, or assets-in-place.
    *   Any existing asset that will generate future cash flows.

*   When building an enterprise DCF (FCFF Model), all future flows are discounted at a constant WACC.
    *   Using a constant WACC, assumes the company manages its capital structure to a target debt-to-value ratio.
    *   What if company plans to change its capital structure significantly?
        *   A company with a high proportion of debt paying it down as cash flow improves.
        *   A constant WACC would overstate the value of the tax shields.
        *   How can we deal with this problem with WACC?
    
---

### The Adjusted Present Value Approach
*   Remember the MM Propositions
    *   In a market with no taxes (among other things), a company's choice of financial structure will not affect the value of its economic assets.
    *   Only market imperfections, such as taxes and distress costs, affect enterprise value.

*   A company in a world with no taxes that has a 50/50 mix of debt and equity.
    *   The company's debt has an expected return of 5 percent,
    
    *   The company's equity has an expected return of 15 percent,
    
    *   Its weighted average cost of capital would be 10 percent. $(0.5 * 5\%) + (0.5 * 15\%) = 2.5\% + 7.5\% = 10\%$.
    
    *   Suppose the company decides to issue more debt, using the proceeds to repurchase shares. What happens to WACC?
        *   In a world with MM assumptions, would a change in capital structure would change the cash flow generated by operations?
        *   Would it change the risk of those cash flows?
        *   Neither the company's enterprise value nor its cost of capital would change.
    
    *   Now suppose we have taxes. What happens to WACC when the company issues more debt and use the proceeds to repurchase shares?

---

### The Adjusted Present Value Approach
*   Rather than model the effect of capital-structure changes in the WACC, APV explicitly measures and values the cash flow effects of financing separately.

*   The APV model separates the value of operations into two components:
    *   The value of operations as if the company were all-equity financed (Base Case Value or $V_U$),
    *   The value of tax shields that arise from debt financing (and potentially other financing side effects).
    $$ \text{Adjusted Present Value} = \text{Enterprise Value as if the Company were All-Equity Financed} + \text{Present Value of Tax Shields} $$
    $$ APV = NPV_U + NPVF $$
    where NPVF is the net present value of the financing side effects.

*   In contrast to WACC approach, the APV approach discounts different cash flows at different rates, using several discount rates.
    *   Discount expected free cash flows from operations (FCFF) at unlevered cost of capital ($r_U$).
    *   Discount expected free cash flows from financing side effects (like tax shields) at a different rate (often cost of debt $r_D$, or $r_U$ depending on risk).

*   Capital structure does not affect the discount rates directly, it affects the expected cash flows from tax savings.

*   APV leads to same result as WACC if WACC is correctly adjusted (e.g., for changing capital structure), but it is easier to apply and less restrictive in many situations, especially with non-constant debt levels.

### The Adjusted Present Value Approach
*   Main side effects (imperfections) creating NPVF:
    *   The tax subsidy to debt (Interest Tax Shields)
    *   The costs of issuing new securities (Flotation Costs)
    *   The costs of financial distress
    *   Subsidies to debt financing
    *   (Other side effects could include hedges, issue costs, etc.)

    **APV = Base-case value ($V_U$) + Value of all financing side effects (NPVF)**

    * Base-case value = value of the project as if it were financed entirely with equity.

    * Value of financing side effects includes PV(interest tax shields), PV(costs of financial distress), PV(subsidies), PV(hedges), PV(issue costs), PV(other costs).

    * APV unbundles components of value and analyzes each one separately.
    * In contrast, WACC bundles all financing side effects into the discount rate.
    
---

### The Adjusted Present Value Approach

*   To build an APV valuation:
    
    **Value the company as if it were all-equity financed ($V^U$).**
        
    *   Discount the Free Cash Flows to the Firm (FCFF) by the unlevered cost of equity ($r_U$).
    
    **Add the Net Present Value of all Financing side effects ($NPVF$).**
        
    *   $NPVF$ primarily includes the Present Value of Interest Tax Shields (PV(ITS)).
        
    *   It can also include the PV of flotation costs (negative), financial distress costs (negative), debt subsidies (positive), etc.

    $$ V^L = V^U + NPVF $$

*   Assuming the primary financing side effect is the interest tax shield, and its value influences the cost of equity:
    $$ r_E = r_U + \frac{D}{E}(r_U - r_D)(1 - \tau_C) $$


*   The cost of equity for a levered firm ($r_E$) equals:
    *   The unlevered cost of equity ($r_U$), reflecting business risk,
    *   plus a premium for the financial risk of leverage ($\frac{D}{E}(r_U - r_D)$),
    *   adjusted downwards by the benefit of debt's tax deductibility (captured by the $(1-\tau_C)$ term in the leverage premium when $r_U > r_D$).

### The Adjusted Present Value Approach

Epsilon Dynamics (Ε), currently an all-equity firm, is considering a new project.
*   Project Initial Investment: $700,000
*   Perpetual Annual Cash Inflows: $800,000
*   Perpetual Annual Cash Costs: $650,000
*   Project's Unlevered Cost of Capital ($r_U$): 18%
*   Corporate Tax Rate ($\tau_C$): 25%

1.  Calculate the Net Present Value of the project if it were financed entirely by equity ($NPV_U$).
2.  Assume Epsilon Dynamics finances the project such that debt amounts to 30% of the project's levered value ($D = 0.30 V^L$).
    *   How much debt ($D$) will the firm use?
    *   What is the Adjusted Present Value (APV) of the project?

---

### The Adjusted Present Value Approach

Initial Investment (I) = \$700,000, Cash Inflows = \$800,000, Cash Costs = \$650,000

Operating Income (EBT) = \$800,000 - \$650,000 = \$150,000 $\quad \Rightarrow \quad$ Tax ($\tau_C=25\%$) = \$150,000 \times 0.25 = \$37,500

NOPAT (Unlevered Cash Flow) = EBT - Tax = \$150,000 - \$37,500 = \$112,500 $\quad \text{and} \quad$ Project $r_U = 18\%$

**NPV of the Unlevered Project:**

$$ V^U = \frac{\text{NOPAT}}{r_U} = \frac{\$112,500}{0.18} \approx \$625,000 \quad \Rightarrow \quad NPV_U = V^U - I = \$625,000 - \$700,000 = -\$75,000 $$

**Project Financed with Debt ($D = 0.30 V^L$):**

$V^L = V^U + PV(\text{Interest Tax Shield}) \quad \text{and} \quad PV(ITS) = \tau_C D \quad \Rightarrow \quad V^L = V^U + \tau_C D \quad \Rightarrow \quad V^L = V^U + \tau_C (0.30 V^L) $

$ V^L = \$625,000 + (0.25 \times 0.30 V^L) \quad \Rightarrow \quad V^L \approx \$675,675.68 $
    
$ D = 0.30 \times V^L = 0.30 \times \$675,675.68 \approx \$202,702.70 $

$ APV = V^L - I \quad \Rightarrow \quad APV = \$675,675.68 - \$700,000 = -\$24,324.32 $

Alternatively, 

$ PV(ITS) = \tau_C \times D = 0.25 \times \$202,702.70 \approx \$50,675.68 $

$ APV = NPV_U + PV(ITS) \quad \Rightarrow \quad APV = -\$75,000 + \$50,675.68 = -\$24,324.32 $

### The Adjusted Present Value Approach

Helios Industries (Η) is considering a new project with the following characteristics:

*   Initial Investment: $40 million
*   Project Life: 5 years
*   Depreciation: Straight-line to zero over 5 years
*   Annual Revenues: $50 million
*   Annual Costs (excluding depreciation): $32 million
*   Corporate Tax Rate ($\tau_C$): 22%
*   Risk-Free Rate ($r_f$): 3.5%
*   Cost of Unlevered Equity ($r_U$): 16%

The project will be partially financed with debt:

*   Net Loan Proceeds desired: $20 million
*   Loan terms: 5-year loan at the risk-free rate of 3.5%, with principal repaid as a bullet payment in year 5.
*   Flotation Costs: 1.5% of the gross loan proceeds, amortized straight-line over the 5-year loan life.

**Calculate the Adjusted Present Value (APV) of this project.**

---

### The Adjusted Present Value Approach

*   Annual Depreciation = $\$40,000,000 / 5 \text{ years} = \$8,000,000$

*   Annual EBIT = Revenues - Costs - Depreciation $\quad \Rightarrow \quad$ EBIT = $\$50,000,000 - \$32,000,000 - \$8,000,000 = \$10,000,000$

*   Annual Unlevered Free Cash Flow ($FCF_U$) for Years 1-5:
    
    $FCF_U = EBIT \times (1 - \tau_C) + \text{Depreciation} \quad \Rightarrow \quad FCF_U = \$10,000,000 \times (1 - 0.22) + \$8,000,000 \quad \Rightarrow \quad FCF_U = \$15,800,000$

*   Present Value of Unlevered FCFs (discounted at $r_U = 16\%$):
    
    $PV(FCF_U) = \$15,800,000 \times \frac{1}{0.16}\left(1 - \frac{1}{(1.16)^5}\right) \quad \Rightarrow \quad PV(FCF_U) = \$15,800,000 \times 3.27429 \approx \$51,733,782$

*   $NPV_U = PV(FCF_U) - \text{Initial Investment} \quad \Rightarrow \quad NPV_U = \$51,733,782 - \$40,000,000 = \$11,733,782$

**Financing Side Effects Valuation**

*   **Flotation Costs & Tax Shield on Flotation Costs:**
    *   Gross Loan Proceeds = Net Proceeds / (1 - Flotation Cost %) $\quad \Rightarrow \quad$ Gross Loan = $\$20,000,000 / (1 - 0.015) \approx \$20,304,569$

    *   Flotation Costs (FC) = Gross Loan - Net Proceeds $\quad \Rightarrow \quad$ FC = $\$20,304,569 - \$20,000,000 = \$304,569$ (Cash outflow at Year 0)
    *   Annual Amortization of Flotation Costs = $\$304,569 / 5 = \$60,913.80

    $\Rightarrow \quad$ Annual Tax Shield from Amortization ($TS_{Flot}$) = Amortization $\times \tau_C $ $\quad \Rightarrow \quad$  $TS_{Flot} = $\$60,913.80 \times 0.22 = \$13,401.04$

    *   PV(TS on Flotation Costs) - discount at $r_f=3.5\%$:
        $PV(TS_{Flot}) = \$13,401.04 \times \frac{1}{0.035}\left(1 - \frac{1}{(1.035)^5}\right) \quad \Rightarrow \quad PV(TS_{Flot}) \approx \$60,510$

*   **Interest Tax Shield (ITS):**
    *   Loan Amount for Interest Calculation (Gross Loan) = $\$20,304,569$
    *   Annual Interest Payment = Gross Loan $\times r_f = \$20,304,569 \times 0.035 = \$710,659.92$
    *   Annual Interest Tax Shield (TS_Int) = Interest Payment $\times \tau_C \quad \Rightarrow \quad TS_{Int} = $\$710,659.92 \times 0.22 = \$156,345.18$
    *   PV(Interest Tax Shield) - discount at $r_f=3.5\%$:
        $PV(ITS) = \$156,345.18 \times \frac{1}{0.035}\left(1 - \frac{1}{(1.035)^5}\right) \quad \Rightarrow \quad PV(ITS) = \approx \$705,997$

*   **Net Present Value of Financing ($NPV_F$):**
$NPV_F = PV(ITS) + PV(TS_{Flot}) - \text{Flotation Costs (at T0)} \quad \Rightarrow \quad NPV_F = \$705,997 + \$60,510 - \$304,569 = \$461,938$

**Adjusted Present Value (APV)**

$APV = NPV_U + NPV_F \quad \Rightarrow \quad APV = \$11,733,782 + \$461,938 = \$12,195,720$

### The Adjusted Present Value Approach

Theta Corp is a rental car company determining whether to add 30 cars to its fleet. The company fully depreciates all rental cars over five years using straight-line method. The new cars are expected to generate $240,000 per year in earnings before taxes and depreciation (EBITDA) for five years. The company is entirely financed by equity with a 30\% tax rate. The required return on unlevered equity ($r_U$) is 12%, and the new fleet will not change company risk. Assume risk-free rate of 7%.

* What is the maximum price Theta should pay for the new fleet if it remains all-equity financed?

* Suppose Theta can purchase the fleet for $720,000. Additionally, assume the company can issue $480,000 of five-year debt to finance the project at the risk-free rate (7%). All principal will be repaid in one balloon payment at the end of the fifth year. What is the adjusted present value (APV) of the project?

---

### The Adjusted Present Value Approach

$$P = PV[(1-\tau_c)(EBITDA)] + PV(\text{Depreciation Tax Shield})$$

After-tax operating cash flow: $(1-0.30) \times 240,000 = 168,000$

$$PV(\text{Operating CF}) = 168,000 \times \frac{1-(1.12)^{-5}}{0.12} = 605,606$$

Depreciation tax shield discounted at risk-free rate:
$$PV(\text{Tax Shield}) = \frac{P}{5} \times 0.30 \times \frac{1-(1.07)^{-5}}{0.07} = \frac{P}{5} \times 0.30 \times 4.1002 = 0.2460P$$

$$P = 605,606 + 0.2460P \quad \Rightarrow \quad P(1-0.2460) = 605,606 \quad \Rightarrow \quad P = \$803,295$$

---
$$APV = NPV(\text{All-Equity}) + NPV(\text{Financing Side Effects})$$

**All-Equity NPV**
$$PV(\text{Tax Shield}) = \frac{720,000}{5} \times 0.30 \times 4.1002 = 144,000 \times 0.30 \times 4.1002 = 177,129 \\ \Rightarrow \quad NPV(\text{All-Equity}) = -720,000 + 605,606 + 177,129 = \$62,735$$

**Financing Side Effects:**

Annual interest = \$480,000 \times 0.07 = 33,600 $\quad \Rightarrow \quad$ After-tax interest = \$33,600 \times (1-0.30) = 23,520$

$$PV(\text{After-tax Interest}) = 23,520 \times 4.1002 = 96,437$$
$$PV(\text{Principal Repayment}) = \frac{480,000}{(1.07)^5} = 342,262$$
$$NPV(\text{Financing}) = 480,000 - 96,437 - 342,262 = \$41,301$$

$$APV = 62,735 + 41,301 = \$104,036$$
