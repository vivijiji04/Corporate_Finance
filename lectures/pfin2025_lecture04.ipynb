### Equity vs. Debt Financing

* **Adding debt to the financial mix**
    * The cost of debt is lower than the cost of equity.
        * Why is debt cheaper than equity?
            * The lender gets first claim on the cash flows and a contractually pre-set cash flow.
            * The equity investor is last in line and demand a higher rate of return than the lender does.

* **Benefits of Debt**
    * Tax Benefits
    * Adds discipline to management

* **Costs of Debt**
    * Bankruptcy Costs
        * The probability of bankruptcy depends upon how uncertain the future cash flows are. (Stable vs. Volatile cash flows)
        * The cost of going bankrupt
            * Direct costs: Legal and other deadweight costs
            * Indirect costs: Costs arising because people perceive you to be in financial trouble
    * Agency Costs
        * Stockholders investing in riskier projects than the lenders would want them to
        * Stockholders paying themselves large dividends the lenders would rather have them keep the cash in the business.
    * Loss of Future Flexibility

### Equity vs. Debt Financing

* **Proposition 1:** Other things being equal, the higher the marginal tax rate of a business, the more debt it will have in its capital structure.

* **Proposition 2:** Firms with more volatile earnings and cash flows will have higher probabilities of bankruptcy at any given level of debt and for any given level of earnings.

* **Proposition 3:** Other things being equal, the greater the indirect bankruptcy cost, the less debt the firm can afford to use for any given level of debt.

* **Proposition 4:** Other things being equal, the greater the agency problems associated with lending to a firm, the less debt the firm can afford to use.

* **Proposition 5:** Other things remaining equal, the more uncertain a firm is about its future financing requirements and projects, the less debt the firm will use for financing current projects.

### Modigliani-Miller Propositions: Leverage, Arbitrage, and Firm Value

* **MM Proposition I**
    * *In a perfect capital market, the total value of a firm's securities is equal to the market value of the total cash flows generated by its assets and is not affected by its choice of capital structure.*

* **MM Proposition II**
    * *The cost of capital of levered equity increases with the firm's market value debt-equity ratio.*

* **Perfect capital markets**
    * Investors and firms can trade the same set of securities at competitive market prices equal to the present value of their future cash flows.
    * There are no taxes, transaction costs, or issuance costs associated with security trading.
    * A firm's financing decisions do not change the cash flows generated by its investments, nor do they reveal new information about them.

* **Remember:** If securities are fairly priced, then buying or selling securities has an NPV of zero and, therefore, should not change the value of a firm.

### Key Costs of Capital

*   **$r_A$ (Cost of Assets / Unlevered Cost of Capital):**
    *   Reflects the inherent business risk** of the firm's operating assets.
    *   The return required by investors if the firm had no debt.
    *   Synonymous with **$r_U$ (Cost of Unlevered Equity)**.

*   **$r_U$ (Cost of Unlevered Equity):**
    *   The required rate of return on an all-equity financed firm.
    *   Driven purely by the firm's business risk.

*   **$r_{WACC}$ (Weighted Average Cost of Capital):**
    *   The firm's blended/average cost of financing** from all sources (debt and equity).
    *   The minimum return a company needs to earn on its asset base to satisfy its creditors and owners.
    *   $r_{WACC} = \frac{E}{V} r_E + \frac{D}{V} r_D$

*   **$r_E$ (Cost of Levered Equity / $r_{levered}$):**
    *   The return required by equity holders in a firm that uses debt.
    *   Reflects both business risk AND financial risk (risk added by leverage).
    *   Increases as the firm takes on more debt.

*   **$r_D$ (Cost of Debt):**
    *   The return required by debt holders.
    *   Generally lower than $r_E$ due to debt's lower risk profile (priority claim).
    *   May increase at very high levels of debt as default risk rises.

### Modigliani-Miller Propositions: Leverage, Arbitrage, and Firm Value

$$E + D = U = A$$

$$\frac{E}{E+D}R_E + \frac{D}{E+D}R_D = R_U$$

* E and D denote the market value of equity and debt if the firm is levered, respectively.
* U is the market value of equity if the firm is unlevered; and A is the market value of the firm's assets.
* R<sub>E</sub>, R<sub>D</sub>, and R<sub>U</sub> are the returns of levered equity, debt, and unlevered equity.

$$R_E = \underbrace{R_U}_{\substack{\text{Risk} \\ \text{without} \\ \text{leverage}}} + \underbrace{\frac{D}{E}(R_U - R_D)}_{\substack{\text{Additional risk} \\ \text{due to leverage}}}$$

**Cost of Capital of Levered Equity**
$$r_E = r_U + \frac{D}{E}(r_U - r_D)$$

### Modigliani-Miller Propositions: Impact of Recapitalization

*   A firm has a market debt-equity ratio of **2**. Its current debt cost of capital is **5%**, and its equity cost of capital is **12.5%**. If the firm issues equity and uses the proceeds to repay its debt and reduce its debt-equity ratio to **1**, it will lower its debt cost of capital to **4.5%**.
    *   With perfect capital markets (no taxes, no transaction costs, etc.), what effect will this transaction have on the firm's equity cost of capital and WACC?
    *   What would happen to the equity cost of capital and WACC if the firm issues even more equity and pays off its debt completely (becomes unlevered)?
    *   How would these alternative capital structures affect the firm's enterprise value?

**Initial State (D/E = 2):**

$$\frac{E}{E+D} = \frac{1}{1 + \frac{D}{E}} = \frac{1}{1+2} = \frac{1}{3} \qquad \text{and} \qquad \frac{D}{E+D} = \frac{\frac{D}{E}}{1 + \frac{D}{E}} = \frac{2}{1+2} = \frac{2}{3}$$

$$r_U = \frac{E}{E+D}r_E + \frac{D}{E+D}r_D \quad \Rightarrow \quad r_U = \left(\frac{1}{3}\right) \times 12.5\% + \left(\frac{2}{3}\right) \times 5\% \quad \Rightarrow \quad r_U = 7.5\%$$

In a world with no taxes: $\qquad r_{WACC} = r_U = r_A = 7.5\%$

### Modigliani-Miller Propositions: Impact of Recapitalization

*   A firm has a market debt-equity ratio of **2**. Its current debt cost of capital is **5%**, and its equity cost of capital is **12.5%**. If the firm issues equity and uses the proceeds to repay its debt and reduce its debt-equity ratio to **1**, it will lower its debt cost of capital to **4.5%**.
    *   With perfect capital markets (no taxes, no transaction costs, etc.), what effect will this transaction have on the firm's equity cost of capital and WACC?
    *   What would happen to the equity cost of capital and WACC if the firm issues even more equity and pays off its debt completely (becomes unlevered)?
    *   How would these alternative capital structures affect the firm's enterprise value?

In a world with no taxes: $\qquad r_{WACC} = r_U = r_A = 7.5\%$

**New State (D/E = 1, $r_D = 4.5\%$):**
Use MM Proposition II (without taxes) to find the new equity cost of capital ($r_E'$):
$$r_E' = r_U + \frac{D}{E}(r_U - r_D') \qquad \Rightarrow \qquad r_E' = 7.5\% + 1 \times (7.5\% - 4.5\%) \qquad \Rightarrow \qquad r_E' = 10.5\%$$

*   The new equity cost of capital is **10.5%**.
*   The WACC remains **7.5%** (equal to $r_U$, as MM Proposition I states firm value, and thus WACC, is unaffected by capital structure in a no-tax world).

**Completely Unlevered State (D/E = 0):**
If the firm pays off all its debt, D/E becomes 0.
$$r_E'' = r_U + \frac{D}{E}(r_U - r_D) \qquad \Rightarrow \qquad r_E'' = 7.5\% + 0 \times (7.5\% - r_D) \qquad \Rightarrow \qquad r_E'' = r_U = 7.5\%$$

*   The equity cost of capital becomes **7.5%**, which is the unlevered cost of capital.
*   The WACC remains **7.5%**.
*   The firm's enterprise value remains unchanged across all these capital structures in perfect capital markets without taxes.

### Modigliani-Miller Propositions: Leverage, Arbitrage, and Firm Value
* **Levered and unlevered betas**
    $$\beta_U = \frac{E}{E+D}\beta_E + \frac{D}{E+D}\beta_D \qquad \Rightarrow \qquad \beta_E = \beta_U + \frac{D}{E}(\beta_U - \beta_D)$$

* When a firm changes its capital structure without changing its investments, 
    * its unlevered beta (asset beta) will remain unaltered. 
    * its equity beta will change to reflect the effect of the capital structure change on its risk.

*   Suppose tech manufacturer **Innovate Corp (INC)** has an equity beta of **1.4** and a debt-equity ratio of **0.25**. Estimate its asset beta (unlevered beta, $\beta_U$) assuming its debt beta is zero ($\beta_D = 0$).

    $$\frac{E}{E+D} = \frac{1}{1 + D/E} = \frac{1}{1 + 0.25} = \frac{1}{1.25} = 0.8 \qquad \text{and} \qquad \frac{D}{E+D} = \frac{D/E}{1 + D/E} = \frac{0.25}{1 + 0.25} = \frac{0.25}{1.25} = 0.2$$

    $$\beta_U = (0.8 \times 1.4) + (0.2 \times 0) \qquad \Rightarrow \qquad \beta_U = 1.12 + 0 = 1.12$$

*   Suppose Innovate Corp were to increase its leverage so that its debt-equity ratio was **0.60**. Assuming its debt beta were still zero ($\beta_D = 0$), what would you expect its equity beta ($\beta_E'$) to be after this increase in leverage?

    $$\beta_E' = \beta_U + \frac{D'}{E'}(\beta_U - \beta_D)$$
    
    $$\beta_E' = 1.12 + 0.60 \times (1.12 - 0) \qquad \Rightarrow \qquad \beta_E' = 1.12 + 0.60 \times 1.12 \qquad \Rightarrow \qquad \beta_E' = 1.792 \approx 1.79$$

### Modigliani-Miller Propositions: Cost of Capital and Beta

*   **Zenith Solutions (ZNS)** has a debt-equity ratio of **0.8**. Its WACC is **10.5%**, and its cost of debt is **5.5%**. What is the company's cost of equity capital ($r_E$)? (Assume no taxes)

    $$r_{WACC} = \frac{E}{E+D}r_E + \frac{D}{E+D}r_D$$
    Given $D/E = 0.8$: $\quad \frac{E}{E+D} = \frac{1}{1 + D/E} = \frac{1}{1 + 0.8} = \frac{1}{1.8} \qquad \text{and} \qquad \frac{D}{E+D} = \frac{D/E}{1 + D/E} = \frac{0.8}{1 + 0.8} = \frac{0.8}{1.8}$

    $$10.5\% = \left(\frac{1}{1.8}\right)r_E + \left(\frac{0.8}{1.8}\right) \times 5.5\% \qquad \Rightarrow \qquad 10.5\% = \frac{1}{1.8}r_E + \frac{4.4\%}{1.8} \qquad \Rightarrow \qquad r_E = 14.5\%$$

*   What is the company's unlevered cost of equity capital ($r_U$)?
    *   In a no-tax world (as per MM assumptions unless stated otherwise), $r_{WACC} = r_U$.
    *   Therefore, $r_U = \mathbf{10.5\%}$.

*   The equity beta of Zenith Solutions is **1.60**, whereas the beta for its debt is **0.20**. What is the company's unlevered beta ($\beta_U$)?
    
    $$\beta_E = \beta_U + \frac{D}{E}(\beta_U - \beta_D)$$
    
    $$1.60 = \beta_U + 0.8 \times (\beta_U - 0.20) \qquad \Rightarrow \qquad \beta_U = \frac{1.76}{1.8} \approx 0.9778 \approx 0.98$$

### Modigliani-Miller Propositions: Levered Equity Beta with Corporate Taxes

**Value of Levered Firm and its Beta:**
   *   From the financing side (claims): $V^L = E + D$.
        The beta of the levered firm's value is the weighted average of the betas of its equity and debt:
        $$ \beta_{V^L} = \frac{E}{V^L}\beta_E + \frac{D}{V^L}\beta_D \quad \quad (1) $$

**Value of Levered Firm and its Beta with Tax Shield):**
   *   From Modigliani-Miller with corporate taxes, the value of a levered firm is the value of an unlevered firm ($V^U$) plus the present value of the interest tax shield:
        $$V^L = V^U + PV(ITS)$$
   *   Assuming perpetual debt and that the risk of the tax shield is equivalent to the risk of the debt ($\beta_{ITS} = \beta_D$), then $PV(ITS) = \tau_C D$. So, 
        $$V^L = V^U + \tau_C D$$
   *   The beta of the levered firm's value can also be seen as a portfolio of the unlevered firm's assets and the tax shield's value contribution:
        $$ V^L \beta_{V^L} = V^U \beta_U + (\tau_C D) \beta_D \quad \Rightarrow \quad \beta_{V^L} = \frac{V^U}{V^L}\beta_U + \frac{\tau_C D}{V^L}\beta_D \quad \quad (2) $$

**Equating the two expressions for $\beta_{V^L}$ and Solving for $\beta_E$:**
   
   From (1) and (2):
   $$ \frac{E}{V^L}\beta_E + \frac{D}{V^L}\beta_D = \frac{V^U}{V^L}\beta_U + \frac{\tau_C D}{V^L}\beta_D \quad \Rightarrow \quad E\beta_E + D\beta_D = V^U\beta_U + \tau_C D\beta_D $$
   Solve for $E\beta_E$:
   $$ E\beta_E = V^U\beta_U + D\beta_D(\tau_C - 1) = V^U\beta_U - D\beta_D(1 - \tau_C) $$
   Substitute $\quad V^U = V^L - \tau_C D = (E+D) - \tau_C D = E + D(1-\tau_C)$:
   $$ \Rightarrow \quad E\beta_E = (E + D(1-\tau_C))\beta_U - D\beta_D(1 - \tau_C) \quad \Rightarrow \quad \beta_E = \frac{E + D(1-\tau_C)}{E}\beta_U - \frac{D(1-\tau_C)}{E}\beta_D $$
   $$ \Rightarrow \quad \beta_E = \left(1 + \frac{D}{E}(1-\tau_C)\right)\beta_U - \frac{D}{E}(1-\tau_C)\beta_D $$
   Rearranging terms:
   $$ \boxed{\beta_E = \beta_U + (\beta_U - \beta_D)\frac{D}{E}(1-\tau_C)} $$

**Special Case: Risk-Free Debt ($\beta_D = 0$) (Hamada Equation):**
    If debt is risk-free, $\beta_D = 0$, the formula simplifies to:
    $$ \boxed{\beta_E = \beta_U\left(1 + (1 - \tau_C)\frac{D}{E}\right)} $$

        * Expected EPS might increase but EPS has become riskier! (Maximizing EPS ignores risk).

### Modigliani-Miller & Financing Decisions: Equity Issuance and Dilution

*   **The "Dilution" Fallacy:** "Issuing equity will dilute existing shareholders' **value per share**, so debt financing should always be preferred." True or False? **(False)**
    *   This statement is a common misconception. 
        * Often overlooks that the cash raised from issuing new shares becomes an asset of the firm, increasing its total value and future earning potential.
    *   While existing shareholders' *percentage* of ownership decreases (ownership dilution), their *value per share* does not necessarily decline.
    *   Critically, if the funds raised are invested in projects with a Net Present Value (NPV) of zero, the firm's total value increases by the exact amount of cash raised. 
        * Assuming new shares are issued at the fair market price, this means the existing **share price remains unchanged**. 
        * While ownership percentage is diluted, the *value* held by existing shareholders (per share) is not diminished by the act of issuing equity for such a project.

### Modigliani-Miller & Financing Decisions: Equity Issuance and Dilution

*   Suppose **NextGen Software Inc.** currently has no debt and **200 million** shares of stock outstanding. These shares are currently trading at **$25** per share. The firm announced an expansion of its operations which will require an investment of **$500 million** that will be financed by issuing new equity. How will the share price change when the new equity is issued today? (Assume the investment itself has an NPV=0).

    *   **Initial Market Value of Equity (and Firm Value, since no debt):**
        
        $200 \text{ million shares} \times \$25/\text{share} = \$5 \text{ billion}$.
    
    *   **Investment Required:** $\$500 \text{ million}$.
    
    *   **Number of New Shares to be Issued:**
        Since the new project has NPV=0, we assume the shares can be issued at the current market price.
        
        $\text{New Shares} = \frac{\text{Investment Required}}{\text{Current Share Price}} = \frac{\$500 \text{ million}}{\$25/\text{share}} = 20 \text{ million shares}$.
    *   **New Total Number of Shares Outstanding:**
        
        $200 \text{ million (old)} + 20 \text{ million (new)} = 220 \text{ million shares}$.
    
    *   **Value of Firm's Assets After Equity Issue and Investment:**
        
        $\text{New Total Assets Value} = \text{Initial Assets Value} + \text{Cash from Issue}\\ \quad \Rightarrow \quad \text{New Total Assets Value} = \$5,000 \text{ million} + \$500 \text{ million} = \$5,500 \text{ million}$.
        
    *   **New Share Price:**
        
        $\text{New Share Price} = \frac{\text{New Total Assets Value (Equity Value)}}{\text{New Total Shares}} = \frac{\$5,500 \text{ million}}{220 \text{ million shares}} = \$25/\text{share}$.

### Modigliani-Miller Propositions: Leverage, Arbitrage, and Firm Value

* **Conservation of value principle for financial markets**
    
    * With perfect capital markets, financial transactions neither add nor destroy value, but instead represent a repackaging of risk (and therefore return).

### Debt and Taxes: Value of the Interest Tax Shield (Alpha Corp.)

*   Suppose **Alpha Corp. (Α)** plans to issue debt. Under the terms of this debt, Alpha Corp. will pay **$30 million** in interest each year for the next **5 years**. In year 5, Alpha Corp. will also repay the principal amount of **$500 million**.
*   These debt payments are considered risk-free, and Alpha Corp.'s marginal tax rate is expected to remain **30%** throughout this period.
*   If the risk-free interest rate is **3%**, by how much does the interest tax shield increase the value of Alpha Corp.?

    $$\text{Annual Interest Tax Shield} = \text{Corporate Tax Rate} \times \text{Annual Interest Payments}$$
    $$\text{Annual Interest Tax Shield} = 30\% \times \$30 \text{ million} = 0.30 \times \$30 \text{ million} = \$9 \text{ million}$$

    $$PV (\text{Annuity}) = \text{Payment} \times \frac{1}{r}\left(1 - \frac{1}{(1+r)^{N}}\right)$$
    $$PV (\text{Interest Tax Shield}) = \$9 \text{ million} \times \frac{1}{0.03}\left(1 - \frac{1}{(1+0.03)^{5}}\right) \quad \Rightarrow \quad PV (\text{Interest Tax Shield}) = \$9 \text{ million} \times 4.5797 \approx \$41.217 \text{ million}$$

*   What is the tax implication of the $500 million principal repayment in year 5?
    *   Any tax shields or tax shield reversals?

*   What are the underlying assumptions when calculating the increase in firm value due to the interest tax shield in this manner?
    *   Sufficient Taxable Income
    *   Constant Tax Rate
    *   Risk of Tax Shields
        * Also assumed that the risk associated with realizing these tax shields is the same as the risk of the debt payments themselves.

### Debt and Taxes
* **The Interest Tax Shield with Permanent Debt**
    
    * Suppose a firm borrows debt (D) and keeps the debt permanently. If the firm's marginal tax rate is $\tau_c$, and if the debt is riskless with a risk-free interest rate $r_f$, then the interest tax shield each year is $\tau_c \times r_f \times D$.
    
    * The present value of this perpetual tax shield, discounted at the risk-free rate $r_f$, is:
    $$PV (Interest Tax Shield) = \frac{\text{Annual Tax Shield}}{r_f} = \frac{\tau_c \times (r_f \times D)}{r_f} = \tau_c \times D$$

* Returning to our previous example and assuming that the debt is permanent (Interest = $100 mio/yr forever from D=$2 bio debt at $r_f=5\%$).
    
    * Annual Tax Shield = $35 \text{ mio}$.
    
    * PV of perpetual tax shield:
    $$ \text{PV(Interest Tax Shield)} = \frac{\text{Annual Tax Shield}}{r_f} = \frac{\$35 \text{ mio}}{0.05} = \$700 \text{ mio} $$
    
    * Alternatively, using the formula for permanent debt:
    $$ \text{PV(Interest Tax Shield)} = \tau_c \times D = 35\% \times \$2 \text{ bio} = 0.35 \times 2000 \text{ mio} = \$700 \text{ mio}$$

### Debt and Taxes
* **WACC with Taxes**
    
    * With tax-deductible interest, the effective after-tax borrowing rate is $r_D(1 - \tau_c)$.
    
    * The Weighted Average Cost of Capital (WACC) formula becomes:
    $$r_{WACC} = \frac{E}{E+D}r_E + \frac{D}{E+D}r_D(1 - \tau_c) \quad \Rightarrow \quad r_{WACC} = \underbrace{\left( \frac{E}{E+D}r_E + \frac{D}{E+D}r_D \right)}_{\text{Pretax WACC}} - \underbrace{\frac{D}{E+D}r_D\tau_c}_{\text{Reduction Due to Interest Tax Shield}}$$

* The firm's pretax WACC is constant since the required return of the firm's investors based on the risk of the firm's assets.
* The (effective after-tax) WACC declines with leverage as the interest tax shield grows.

### Debt and Taxes: Valuing the Interest Tax Shield
*   **Omega Solutions (Ω)** expects FCF of **$7.5 million** next year, growing at **3%** annually.
    *   $r_E = 12\%$, $r_D = 5\%$, Corporate Tax Rate $\tau_C = 25\%$. Target Debt-Equity Ratio (D/E) = **0.60**.
*   If Omega Solutions maintains this D/E ratio, what is the value of its interest tax shield?
    $$\frac{E}{V} = \frac{1}{1 + D/E} = \frac{1}{1.60} \quad \text{and} \quad \frac{D}{V} = \frac{D/E}{1 + D/E} = \frac{0.60}{1.60}$$
    
    $$r_U = \left(\frac{1}{1.60}\right) \times 12\% + \left(\frac{0.60}{1.60}\right) \times 5\% = 7.5\% + 1.875\% = 9.375\%$$

    $$V^U = \frac{FCF_1}{r_U - g} = \frac{\$7.5 \text{ M}}{0.09375 - 0.03} = \frac{\$7.5 \text{ M}}{0.06375} \approx \$117.65 \text{ M}$$

*   **After-Tax WACC ($r_{WACC}$):**
    $$r_{WACC} = \left(\frac{1}{1.60}\right) \times 12\% + \left(\frac{0.60}{1.60}\right) \times 5\% \times (1 - 0.25) \quad \Rightarrow \quad r_{WACC} = 8.90625\%$$

*   **Levered Value ($V^L$):**
    $$V^L = \frac{FCF_1}{r_{WACC} - g} = \frac{\$7.5 \text{ M}}{0.0890625 - 0.03} \approx \$126.98 \text{ M}$$

*   **PV of Interest Tax Shield ($PV(ITS)$):**
    $$\text{PV(Interest Tax Shield)} = V^L - V^U \quad \Rightarrow \quad \text{PV(Interest Tax Shield)} = \$126.98 \text{ M} - \$117.65 \text{ M} \approx \$9.33 \text{ M}$$

### Debt and Taxes: Recapitalization of Gamma Corp. (Γ) - Part 2

*   Initial Value (Unlevered, $V^U$): $500 million
*   PV of Tax Shield: $PV(ITS) = \tau_C \times D = 30\% \times \$150 \text{ million} = \$45 \text{ million}$
*   New Firm Value (Levered, $V^L$): $V^L = V^U + PV(ITS) = \$500m + \$45m = \$545 \text{ million}$
*   Value of Equity after recap (for remaining shareholders): $E = V^L - D = \$545m - \$150m = \$395 \text{ million}$
*   *Shareholder wealth increases by the PV of the tax shield: **$45 million**.*

*   **The Share Repurchase:** The firm will use the $150 million from debt to repurchase shares.
*   **Can Gamma Corp. repurchase shares at the current price of $12.50?**
    *   No. If the recapitalization is announced and the market is efficient, the value of the firm will immediately increase to $V^L = \$545 \text{ million}$ due to the PV of the tax shield.
    *   The share price should rise *before* the repurchase to reflect this new total firm value.
    *   New "cum-rights" price (before shares are repurchased, but after value increase is known):
        $P_{New, PreRep} = V^L / \text{Initial Shares} = \$545 \text{ million} / 40 \text{ million shares} = \$13.625 \text{ per share}$.

*   **Repurchase at the New Equilibrium Price:** Assume the repurchase occurs at $P_R = \$13.625$.
    *   Shares Repurchased ($N_{Rep}$):
        $N_{Rep} = \frac{\text{Cash for Repurchase}}{P_R} = \frac{\$150 \text{ million}}{\$13.625/\text{share}} \approx 11.009 \text{ million shares}$
    *   Shares Remaining Outstanding ($N_{Post}$):
        $N_{Post} = \text{Initial Shares} - N_{Rep} = 40 \text{ million} - 11.009 \text{ million} \approx 28.991 \text{ million shares}$
    *   Value of Equity *after* cash is paid out for repurchase: $E_{Post} = \$395 \text{ million}$.
    *   **New Share Price ($P_{Post}$):**
        $P_{Post} = \frac{E_{Post}}{N_{Post}} = \frac{\$395 \text{ million}}{28.991 \text{ million shares}} \approx \$13.625 \text{ per share}$.

### Personal Taxes and Effective Tax Advantage of Debt
* Interest payments received from debt are usually taxed as ordinary income.
* Equity investors also pay taxes on dividends and capital gains.
* The value of a firm is equal to the amount of money the firm can raise by issuing securities.
    * The amount of money an investor will pay for a security ultimately depends on the benefits the investor will receive.
    * Personal taxes reduce the cash flows to investors and diminish firm value.
    * The actual interest tax shield depends on the reduction in the total taxes (both corporate and personal) that are paid.

|                        |         After-Tax Cash Flows                              |
| :--------------------- | :------------------------------------ |
| To debt holders        | $(1 - \tau_i)$                        |
| To equity holders      | $(1 - \tau_c)(1 - \tau_e)$            |

<br>

$$ (1 - \tau^*)(1 - \tau_i) = (1 - \tau_c)(1 - \tau_e) $$

<br>

If the corporation paid $(1 - \tau^*)$ in interest, debt holders would receive the same amount after taxes as equity holders would receive if the firm paid $1 in profits to equity holders.

<br>

**Effective Tax Advantage of Debt**

$$ \tau^* = 1 - \frac{(1 - \tau_c)(1 - \tau_e)}{(1 - \tau_i)} $$


---

### Personal Taxes and Effective Tax Advantage of Debt

Consider two scenarios for a company and its investors:

**Scenario A:**
   *   Corporate Tax Rate ($\tau_c$) = 30%
   *   Personal Tax Rate on Interest Income ($\tau_i$) = 35%
   *   Effective Personal Tax Rate on Equity Income ($\tau_e$) = 15%

**Scenario B:**
   *   Corporate Tax Rate ($\tau_c$) = 20%
   *   Personal Tax Rate on Interest Income ($\tau_i$) = 25%
   *   Effective Personal Tax Rate on Equity Income ($\tau_e$) = 25%

**Calculate the Effective Tax Advantage of Debt ($\tau^*$) for both scenarios. What does a negative $\tau^*$ imply?**

**Scenario A:**
$$ \tau^*_A = 1 - \frac{(1 - 0.30)(1 - 0.15)}{(1 - 0.35)} \quad \Rightarrow \quad \tau^*_A = 1 - 0.91538 \approx 0.08462 $$

For every $1 of pretax earnings paid out as interest instead of equity income, there is a net tax saving of approximately 8.46 cents across all parties (corporation and investors combined). Debt financing offers a positive effective tax advantage.

**Scenario B:**
$$ \tau^*_B = 1 - \frac{(1 - 0.20)(1 - 0.25)}{(1 - 0.25)} \quad \Rightarrow \quad \tau^*_B = 1 - 0.80 = 0.20 \quad \Rightarrow \quad \tau^*_B = 20\%$$

For every \$1 of pretax earnings paid out as interest instead of equity income, there is a net tax saving of 20 cents. In this case, because the personal tax rates on equity and interest income are the same, the effective tax advantage of debt is simply equal to the corporate tax rate ($\tau^*_B = \tau_c$).

---

### Personal Taxes: Interpreting the Effective Tax Advantage of Debt ($\tau^*$)

**What does a negative $\tau^*$ imply?**

A negative $\tau^*$ indicates that, considering all taxes (corporate and personal), distributing \$1 of earnings as interest results in a higher total tax burden than distributing it as equity income.

This occurs when the term $\frac{(1 - \tau_c)(1 - \tau_e)}{(1 - \tau_i)}$ is greater than 1, which means: 

$$\quad (1 - \tau_c)(1 - \tau_e) > (1 - \tau_i) \quad \Rightarrow \quad (1 - \tau_c) > \frac {(1 - \tau_i)} {(1 - \tau_e)}$$


**Key Conditions Leading to a Negative $\tau^*$:**

1.  High Personal Tax on Interest ($\tau_i$): The personal tax rate on interest income is substantially higher than...
2.  Low Personal Tax on Equity ($\tau_e$): ...the effective personal tax rate on equity income (dividends/capital gains).
3.  Offsetting Corporate Benefit: This disparity in personal tax rates is significant enough to outweigh the benefit of the corporate tax deductibility of interest (represented by $\tau_c$).
