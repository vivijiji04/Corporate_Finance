### Relative Valuation
* In discounted cash flow valuation, we are attempting to estimate the intrinsic value of an asset based upon its capacity to generate cash flows in the future.

* In relative valuation, we are making a judgment on how much an asset is worth by looking at what the market is paying for similar assets.
    
    * If the market is correct, on average, in the way it prices assets, discounted cash flow and relative valuations may converge.
    
    * If the market is systematically over pricing or under pricing a group of assets or an entire sector, discounted cash flow valuations can deviate from relative valuations.

* Most equity research reports are based upon multiples.

* Discounted cash flow techniques are more common in acquisitions and corporate finance.
    
---

### Relative Valuation
* Essential steps in relative valuation
    * Finding comparable assets that are priced by the market.
    * Scaling the market prices to a common variable to generate standardized prices.
    * Adjusting for differences across assets when comparing their standardized values.

* Ensure that the multiple is defined consistently and that it is measured uniformly across the firms being compared.
* Be aware of the cross-sectional distribution of the multiple, not only across firms in the sector being analyzed but also across the entire market.
* Analyze the multiple and understand not only what fundamentals determine the multiple but also how changes in these fundamentals translate into changes in the multiple.
* Finding the right firms to use for comparison and controlling for differences that may persist across these firms.
---

### Relative Valuation
* Consistency
    * Every multiple has a numerator and a denominator.
        * The numerator can be either an equity value (market price, value of equity) or a firm value (enterprise value)
        * The denominator can be an equity measure (EPS, net income, book value of equity) or a firm measure (such as operating income, EBITDA or book value of capital).
    * The numerator and denominator should be defined consistently.
        * If the numerator for a multiple is an equity value, then the denominator should be an equity value as well.
        * If the numerator is a firm value, then the denominator should be a firm value as well.

### Relative Valuation
* Comparable firms
    * A comparable firm is one with cash flows, growth potential, and risk similar to the firm being valued.
    * Is being in the same industry sufficient for companies to be comparable?
        * Being identical in terms of cash flows, growth, and risk is more important than being in the same industry.
            * Look for firms that are similar in terms of valuation fundamentals.
            * Consider all firms in the market as comparable firms and control for differences on the fundamentals across these firms.

### Relative Valuation
* It is important to understand what the key variables are that determine and drive the multipliers.

* Variables that drive discounted cash flow valuation, such as growth, risk, and cash flow models, also affect multipliers.

* Information about the underlying factors affecting a multiplier is important.

**Forward P/E Ratio**

$D_1 = p \times EPS_1$.

$ P_0 = \frac{p \times EPS_1}{r-g} \quad \Rightarrow \quad \frac{P_0}{EPS_1} = \frac{p}{r-g} \quad \Rightarrow \quad\frac{P_0}{E_1} = \frac{\text{Payout Ratio}}{r-g} $

**Trailing P/E Ratio**

$D_1 = p \times EPS_1 = p \times EPS_0 (1+g)$.

$ P_0 = \frac{p \times EPS_0 (1+g)}{r-g} \quad \Rightarrow \quad \frac{P_0}{EPS_0} = \frac{p \times (1+g)}{r-g} \quad \Rightarrow \quad \frac{P_0}{E_0} = \frac{\text{Payout Ratio} \times (1+g)}{r-g} $

### Relative Valuation
* The PEG ratio is the ratio of the P/E ratio to a firm's expected growth rate.
    
    * Widely used particularly for analyzing high-growth firms.
    
    * Correcting the P/E ratio with the growth rate makes it more meaningful to compare firms with different growth rates.

* The expected growth rate of AGE Corp is 20%. The average PEG ratio within its industry is 50. What should be the P/E ratio of AGE Corp?
    
    PEG= PE / g
    
    $50 = \frac{(PE)}{0.20} \quad \rightarrow \quad PE= 50 \times 0.20 \quad \rightarrow \quad PE =10$

* AGE Corp has a P/E ratio of 8 and an expected growth rate of 20%. The average PEG and P/E ratios within its industry are 40 and 12, respectively. How is its valuation based on industry averages?
    
    $PEG= \frac{PE}{g} \quad \rightarrow \quad PEG = \frac{8}{0.20} \quad \rightarrow \quad PEG = 40$

### Relative Valuation

* The book value is equal to the net assets of a firm. It refers to the amount remaining if the company liquidates all its assets and repays all its liabilities.
* A high MV/BV ratio is typical for firms for which the market value is mostly driven by growth prospects.
* A low MV/BV ratio, on the other hand, is typical for "value" firms for which market value is mostly determined by existing assets.
* Empirical studies suggest that the MV/BV ratio is a significant determinant of stock prices.


$$
P_0 = \frac{D_1}{r-g}
$$

$$
P_0 = \frac{EPS_1 \times Payout \, Ratio}{r-g}
$$

$$
ROE = \frac{EPS_1}{BV_0} \quad \text{(where } BV_0 \text{ is Book Value of Equity per share at } t=0 \text{)}
$$

$$
P_0 = \frac{ROE \times BV \times Payout \, Ratio}{r-g}
$$

$$
\frac{MV}{BV} = \frac{ROE \times Payout \, Ratio}{r-g}
$$

$$
g = (1 - p) \times ROE \implies \frac{g}{ROE} = (1-p) \implies p = \frac{ROE - g}{ROE}
$$

$$
\frac{MV}{BV} = \frac{(ROE-g)}{r-g}
$$

### Relative Valuation: Enterprise Value Multiples
* Enterprise value to EBIT, EBITDA
    
* While firms in the same industry may be otherwise comparable, if these firms have different degrees of leverage, they are likely to have different P/E ratios.
    
* Because firm value includes both debt and equity, leverage has less of an impact on the EV/EBITDA and EV/EBIT ratios.
    
* Far fewer firms have negative EV/EBITDA or EV/EBIT ratios.
    
* The important variables that determine the level of EV/EBITDA and EV/EBIT ratios are,
    * Growth rate, risk, and return on capital.
    
    * Firms with higher growth, lower risk (cost of capital), and higher return on capital should trade at higher EBITDA and EBIT multiples.

---

### Relative Valuation: Enterprise Value Multiples
* All other variables being equal, an increase in the expected growth rate of operating income will increase the EV multipliers.
    
* When comparing EV multipliers among firms with widely distributed growth rates, growth rate differences should be explicitly controlled. Otherwise, the results obtained will be biased.
        
    * It may be deduced that companies with lower growth are undervalued and companies with higher growth are overvalued.
       
    * Other variables being constant, high returns with high growth, in the long run, will increase EV multipliers. Thus, firms with stronger and more sustainable competitive advantages are expected to have higher firm value multipliers than firms without these advantages.

### Relative Valuation: Enterprise Value Multiples

Adele, a tech firm, has earnings per share ($EPS$) of $3.20 and EBITDA of $45.0 million. Adele also has 8 million shares outstanding and debt of $180 million (net of cash). You believe Beyoncé, a comparable firm, has a P/E multiple of 15.0 and an EV/EBITDA multiple of 9. Estimate the value of Adele's shares using both multiples. Which estimate is likely to be more accurate?
    

**Share Price using P/E Multiple:**

$ P_0^{P/E} = \text{Comparable P/E} \times EPS \quad \Rightarrow \quad P_0^{P/E} = 15.0 \times \$3.20 = \$48.00 $

**Share Price using EV/EBITDA Multiple:**

$ EV = \text{Comparable EV/EBITDA} \times EBITDA \quad \Rightarrow \quad EV = 9.0 \times \$45.0M = \$405.0M $

$ \text{Equity Value (E)} = EV - \text{Debt (net of cash)} \quad \Rightarrow \quad E = \$405.0M - \$180M = \$225.0M $

$ P_0^{EV/EBITDA} = \frac{E}{\text{Shares Outstanding}} \quad \Rightarrow \quad P_0^{EV/EBITDA} = \frac{\$225.0M}{8M \text{ shares}} = \$28.125 $

### Relative Valuation: Enterprise Value Multiples

Taylor, a tech firm specializing in cloud-based software solutions, has current earnings per share ($EPS$) of $1.35 and Earnings Before Interest, Taxes, Depreciation, and Amortization (EBITDA) of $25 million. Taylor has 10 million shares outstanding and net debt (debt less cash) of $50 million. Its depreciation expense is $5 million, and interest expense is $2 million. The corporate tax rate for both firms is 25%.

You are valuing Taylor using Mariah, a comparable tech firm specializing in cloud infrastructure and hardware. Mariah has $EPS$ of $1.25 and EBITDA of $40 million. Mariah's depreciation expense is $15 million, and interest expense is $5 million. Mariah has 12 million shares outstanding and net debt of $100 million. Mariah's current stock price is $30.00.


**Mariah's P/E multiple:**  $\quad P/E = P_0 / EPS = \$30.00 / \$1.25 = 24.0$

**Mariah's EV/EBITDA multiple:**
    
Equity Value ($E$) = $P_0 \times N = \$30.00 \times 12M = \$360M$
    
Enterprise Value ($EV$) = $E + \text{Net Debt} = \$360M + \$100M = \$460M$
    
$EV/EBITDA = EV / EBITDA = \$460M / \$40M = 11.5$

**Taylor's Share Price using Multiples from Mariah:**

*   **Using P/E Multiple:**
    $P_0^{P/E} = \text{Mariah's P/E} \times \text{Taylor's EPS} \quad \Rightarrow \quad P_0^{P/E} = 24.0 \times \$1.35 = \$32.40$

*   **Using EV/EBITDA Multiple:**
    $EV^{\text{Taylor}} = \text{Mariah's EV/EBITDA} \times \text{Taylor's EBITDA} \quad \Rightarrow \quad EV^{\text{Taylor}} = 11.5 \times \$25M = \$287.5M$

    $\text{Equity Value}^{\text{Taylor}} = EV^{\text{Taylor}} - \text{Taylor's Net Debt} \quad \Rightarrow \quad \text{Equity Value}^{\text{Taylor}} = \$287.5M - \$50M = \$237.5M$
    
    $P_0^{EV/EBITDA} = \frac{\text{Equity Value}^{\text{Taylor}}}{\text{Taylor's Shares Outstanding}} \quad \Rightarrow \quad P_0^{EV/EBITDA} = \frac{\$237.5M}{10M \text{ shares}} = \$23.75$

---

### Relative Valuation: Enterprise Value Multiples

*   **Impact of Depreciation:**
    *   Taylor has lower depreciation ($5M$) relative to its EBITDA ($25M$) than Mariah ($15M$ depreciation relative to $40M$ EBITDA). This suggests Taylor might be a less fixed-asset-intensive business (e.g., software services) compared to Mariah (e.g., cloud infrastructure/hardware).
    *   P/E multiples are based on Net Income, which is after depreciation. Differences in depreciation across comparable firms can significantly distort Net Income and thus EPS.
    *   EV/EBITDA multiples, by using EBITDA, are unaffected by such differences in asset intensity and depreciation policies, making them more suitable for comparing firms with varying capital expenditure histories or asset bases.

*   **Impact of Capital Structure:**
    *   P/E multiples are also affected by capital structure differences.
    *   EV/EBITDA, by valuing the entire firm (Enterprise Value), is less sensitive to the mix of debt and equity financing. It provides a cleaner comparison of the core operating business value before financing effects are considered.

```python
import numpy as np
import matplotlib.pyplot as plt
import ipywidgets as widgets
from ipywidgets import HBox, VBox, Layout, Label
from IPython.display import display, clear_output

# --- Fixed Parameters for Taylor (The Company Being Valued) ---
TAYLOR_EBITDA_BASE = 25 # $ million
TAYLOR_SHARES_OUTSTANDING = 10 # million shares (default starting value)

# --- Fixed Parameters for Mariah (The Comparable Company) ---
MARIAH_PE_MULTIPLE = 24.0 # P/E from Mariah's P0/EPS
MARIAH_EV_EBITDA_MULTIPLE = 11.5 # EV/EBITDA from Mariah

# --- Output widget to display the dynamic plot and numerical prints ---
plot_output = widgets.Output()

# --- Define the Interactive Plotting and Calculation Function ---
def update_valuation_plot(depreciation_val, cost_of_debt_val, tax_rate_val, net_debt_val, shares_outstanding_val):
    with plot_output:
        clear_output(wait=True)

        if shares_outstanding_val < 0.1:
            shares_outstanding_val = 0.1
        
        # Calculate Interest Expense based on Net Debt and Cost of Debt
        interest_expense_calc = net_debt_val * cost_of_debt_val
        
        ebit = TAYLOR_EBITDA_BASE - depreciation_val
        ebt = ebit - interest_expense_calc # Use calculated interest expense
        
        taxes = np.maximum(0, ebt * tax_rate_val)
        
        net_income = ebt - taxes
        eps = net_income / shares_outstanding_val if shares_outstanding_val != 0 else np.nan

        p_pe_estimate = MARIAH_PE_MULTIPLE * eps
        
        ev_ebitda_estimate = MARIAH_EV_EBITDA_MULTIPLE * TAYLOR_EBITDA_BASE
        equity_value_from_ev = ev_ebitda_estimate - net_debt_val
        p_ev_ebitda_estimate = equity_value_from_ev / shares_outstanding_val if shares_outstanding_val != 0 else np.nan

        # --- Visualization ---
        fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(15, 7))
        fig.suptitle("Taylor's Valuation: Impact of Accounting & Capital Structure", fontsize=16)

        x_label = ['Taylor']
        bar_width = 0.5
        x_pos = np.arange(len(x_label))

        bottom_ni = net_income
        bottom_taxes = bottom_ni + taxes
        bottom_interest = bottom_taxes + interest_expense_calc # Use calculated
        bottom_depreciation = bottom_interest + depreciation_val

        ax1.bar(x_pos, bottom_ni, bar_width, label='Net Income', color='skyblue', edgecolor='black')
        ax1.bar(x_pos, taxes, bar_width, bottom=bottom_ni, label='Taxes', color='lightcoral', edgecolor='black')
        ax1.bar(x_pos, interest_expense_calc, bar_width, bottom=bottom_taxes, label='Interest Expense', color='lightgreen', edgecolor='black') # Use calculated
        ax1.bar(x_pos, depreciation_val, bar_width, bottom=bottom_interest, label='Depreciation', color='gold', edgecolor='black')

        ax1.axhline(TAYLOR_EBITDA_BASE, color='gray', linestyle='--', linewidth=2, label='EBITDA Level')
        ax1.text(x_pos[0] + bar_width, TAYLOR_EBITDA_BASE + 0.5, f'EBITDA: ${TAYLOR_EBITDA_BASE:.1f}M', va='bottom', fontsize=10, color='gray')

        ax1.set_ylabel('Amount ($ millions)', fontsize=12)
        ax1.set_xticks(x_pos)
        ax1.set_xticklabels(x_label, fontsize=12)
        ax1.set_title('Income Statement Breakdown', fontsize=14)
        ax1.legend(loc='upper left', bbox_to_anchor=(1, 1))
        ax1.grid(axis='y', linestyle='--', alpha=0.7)
        
        min_y = min(0, np.floor(net_income / 5) * 5 if net_income < 0 else 0)
        max_y = TAYLOR_EBITDA_BASE * 1.2
        ax1.set_ylim(min_y - 5, max_y + 5)


        metrics = ['P/E Based Price', 'EV/EBITDA Based Price']
        prices = [p_pe_estimate, p_ev_ebitda_estimate]

        prices = np.nan_to_num(prices, nan=0.0, posinf=500, neginf=-500)
        prices = np.clip(prices, -500, 500)

        ax2.bar(metrics, prices, color=['darkred', 'darkblue'], edgecolor='black', width=0.6)

        for i, price in enumerate(prices):
            ax2.text(i, price + np.sign(price)*0.5 + (0.5 if price==0 else 0), f'${price:.2f}', ha='center', va='bottom' if price >= 0 else 'top', fontsize=11, color='black')

        ax2.set_ylabel('Share Price ($)', fontsize=12)
        ax2.set_title('Estimated Share Price per Method', fontsize=14)
        ax2.grid(axis='y', linestyle='--', alpha=0.7)
        ax2.axhline(0, color='black', linewidth=0.8)

        min_p_y = min(0, np.floor(np.min(prices) / 10) * 10 if np.min(prices) < 0 else 0)
        max_p_y = np.ceil(np.max(prices) / 10) * 10 if np.max(prices) > 0 else 10
        ax2.set_ylim(min_p_y - 5, max_p_y + 5)


        plt.tight_layout(rect=[0, 0.03, 1, 0.95])
        plt.show()

        # --- Print Numerical Results Below Plot ---
        #print("\n--- Current Financial Metrics & Valuation Estimates for Taylor ---")
        #print(f"Base EBITDA: ${TAYLOR_EBITDA_BASE:.1f}M")
        #print(f"Depreciation: ${depreciation_val:.1f}M")
        #print(f"Net Debt: ${net_debt_val:.1f}M")
        #print(f"Cost of Debt (rD): {cost_of_debt_val:.1%}")
        #print(f"Calculated Interest Expense: ${interest_expense_calc:.2f}M")
        #print(f"Tax Rate: {tax_rate_val:.0%}")
        #print(f"Shares Outstanding: {shares_outstanding_val:.1f}M")
        #print("--------------------------------------------------")
        #print(f"Calculated Net Income (NI): ${net_income:.2f}M")
        #print(f"Calculated Earnings Per Share (EPS): ${eps:.2f}")
        #print("--------------------------------------------------")
        #print(f"Estimated Share Price (P/E Based): ${p_pe_estimate:.2f}")
        #print(f"Estimated Share Price (EV/EBITDA Based): ${p_ev_ebitda_estimate:.2f}")
        #print("\n**Observation:**")
        #print("Adjust 'Depreciation', 'Net Debt', and 'Cost of Debt' to see their impact.")
        #print("P/E based price is sensitive to D/A, Net Debt (via interest), and Cost of Debt (via interest).")
        #print("EV/EBITDA based price is stable against D/A and Cost of Debt, but changes with Net Debt (as it's subtracted from EV to get Equity Value).")


# --- Create Interactive Widgets ---
slider_layout = Layout(width='200px')
label_layout = Layout(width='180px') # Increased label width
style = {'description_width': 'initial'}

depreciation_widget = widgets.FloatSlider(value=5, min=0, max=20, step=1, readout_format='.1f', layout=slider_layout)
cost_of_debt_widget = widgets.FloatSlider(value=0.05, min=0.01, max=0.15, step=0.005, readout_format='.1%', layout=slider_layout) 
tax_rate_widget = widgets.FloatSlider(value=0.25, min=0, max=0.5, step=0.01, readout_format='.0%', layout=slider_layout)
net_debt_widget = widgets.FloatSlider(value=50, min=0, max=200, step=10, readout_format='.1f', layout=slider_layout)
shares_widget = widgets.FloatSlider(value=10, min=1, max=20, step=1, readout_format='.1f', layout=slider_layout)

# layout
left_column_sliders = VBox([
    HBox([Label('Depreciation ($M):', layout=label_layout), depreciation_widget]),
    HBox([Label('Net Debt ($M):', layout=label_layout), net_debt_widget]),
    HBox([Label('Cost of Debt (rD):', layout=label_layout), cost_of_debt_widget])
])

right_column_sliders = VBox([
    HBox([Label('Tax Rate (τC):', layout=label_layout), tax_rate_widget]),
    HBox([Label('Shares Outstanding (M):', layout=label_layout), shares_widget])
])

controls_combined = HBox([left_column_sliders, right_column_sliders])
controls_combined.layout = Layout(justify_content='space-around', align_items='flex-start', width='100%') # Ensure controls use full width

info_text_widgets = VBox([
    Label(f"Base EBITDA for Taylor: ${TAYLOR_EBITDA_BASE}M", style=style),
    Label(f"Mariah's P/E Multiple: {MARIAH_PE_MULTIPLE:.1f}", style=style),
    Label(f"Mariah's EV/EBITDA Multiple: {MARIAH_EV_EBITDA_MULTIPLE:.1f}", style=style),
    Label("Adjust the sliders below to see the impact on Taylor's valuation:", style=style)
])

interactive_plot_connection = widgets.interactive_output(
    update_valuation_plot, {
        'depreciation_val': depreciation_widget,
        'cost_of_debt_val': cost_of_debt_widget, 
        'tax_rate_val': tax_rate_widget,
        'net_debt_val': net_debt_widget,
        'shares_outstanding_val': shares_widget
    }
)

display(info_text_widgets, controls_combined, plot_output)
```

LUMI's per share data: EPS = €1.25; CFPS = €2.80; BVPS = €8.50; SPS = €15.70

*   Value based on P/E: $\quad P_{LUMI, P/E} = \text{Mean P/E} \times \text{LUMI EPS} = 21.6667 \times €1.25 \approx €27.08$

*   Value based on P/CF: $\quad P_{LUMI, P/CF} = \text{Mean P/CF} \times \text{LUMI CFPS} = 9.9167 \times €2.80 \approx €27.77$

*   Value based on P/BV: $\quad  P_{LUMI, P/BV} = \text{Mean P/BV} \times \text{LUMI BVPS} = 2.8111 \times €8.50 \approx €23.89$

*   Value based on P/S: $\quad P_{LUMI, P/S} = \text{Mean P/S} \times \text{LUMI SPS} = 1.8222 \times €15.70 \approx €28.63$

**Preliminary Valuation for LUMI:**

$ P_{LUMI, Avg} = \frac{P_{LUMI, P/E} + P_{LUMI, P/CF} + P_{LUMI, P/BV} + P_{LUMI, P/S}}{4} $

$ P_{LUMI, Avg} = \frac{€27.08 + €27.77 + €23.89 + €28.63}{4} \quad \Rightarrow \quad P_{LUMI, Avg} = \frac{€107.37}{4} \approx €26.84 $

From the prior analysis using comparable company multiples, the preliminary estimated standalone per-share value for LUMI was determined to be approximately €26.84. **Comparable Takeover Data:**

| :--------------- | --------------------------------: | ------------------------: |

| Urban Nest       | 18.75                             | 23.81                     |

3.  Using LUMI's preliminary estimated standalone stock value (€26.84) and the mean takeover premium, estimate a potential takeover price per share for LUMI.

### Relative Valuation

**Takeover Premium Percentage for Each Comparable Acquisition:**

Takeover Premium (%) = $\left( \frac{\text{Actual Takeover Price}}{\text{Stock Price Prior to Takeover}} - 1 \right) \times 100\%$

| Acquired Company | Price Prior (€) (a) | Takeover Price (€) (b) | Premium = (b/a - 1) | Premium (%) |
| :--------------- | ------------------: | ---------------------: | ------------------: | ----------: |
| Lifestyle Trendz | 22.50               | 28.13                  | $28.13/22.50 - 1 \approx 0.2502$ | $25.02\%$   |
| Nordic Chic      | 30.00               | 38.25                  | $38.25/30.00 - 1 = 0.2750$     | $27.50\%$   |
| Urban Nest       | 18.75               | 23.81                  | $23.81/18.75 - 1 \approx 0.2699$ | $26.99\%$   |
| Design Direct    | 25.20               | 31.50                  | $31.50/25.20 - 1 = 0.2500$     | $25.00\%$   |

**Mean Takeover Premium:**

$ \text{Mean Premium} = \frac{25.02\% + 27.50\% + 26.99\% + 25.00\%}{4} \quad \Rightarrow \quad \text{Mean Premium} = \frac{104.51\%}{4} \approx 26.1275\% \approx 26.13\% $

**Potential Takeover Price for LUMI:**

LUMI's Preliminary Estimated Standalone Stock Value = €26.84

Mean Takeover Premium = 26.13%

$ \text{Estimated Takeover Price for LUMI} = \text{LUMI Estimated Value} \times (1 + \text{Mean Premium}) $

$ \text{Estimated Takeover Price for LUMI} = €26.84 \times (1 + 0.2613) \quad \Rightarrow \quad \text{Estimated Takeover Price for LUMI} \approx €33.85 $
